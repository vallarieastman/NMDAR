    Introduction:
  </h3>
<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    The crystal structure of the NR1 ligand-binding domain (LBD) provides a detailed molecular framework for understanding how glycine binding drives NMDA receptor activation. Furukawa & Gouaux (2003) describe the structural organization of the NR1 ligand-binding domain (LBD) as composed of two discontinuous polypeptide segments: S1 (N-terminal portion) forming lobe D1 and S2 (C-terminal portion) forming lobe D2. Together, they fold into a hinged bilobed clamshell typical of ionotri=opic glutamate receptor LBDs. Upon agonist (glycine) binding, the D2 lobe rotates and moves toward D1, resulting in cleft closure that pulls on the transmembrane gating linkers, coupling ligand binding to pore opening (Furukawa & Gouaux, 2003). This cleft closure represents a key conformational change that transmits mechanical force via the S2–M3 linker to open the ion channel gate (Furukawa & Gouaux, 2003). Furukawa & Gouaux (2003) used the crystal structure PDB 1PBQ and noted that key residues forming the binding cleft include: D1 residues (Thr518, Arg523) form the rigid anchoring surface, holding glycine in place, D2 residues (Ser688, Thr690, Glu712, Asp732) form the mobile “closing” lobe, which moves toward D1 when glycine binds, and the resulting D1–D2 cleft closure is the mechanical trigger that initiates channel gating through tension on the S2–M3 linker (Figure 1). </p>

When the D1–D2 cleft closes upon glycine binding, the D2 lobe moves toward D1, pulling on this GT linker. This tension is transmitted to the M3 helix, triggering channel opening. The disulfide (S–S) bonds stabilize the LBD’s tertiary structure and maintain the correct orientation of the D1 and D2 lobes. There are two key disulfide pairs: C420–C454 and C436–C455 (in D1) and C744–C798 (in D2). Loop 1 (in Domain 1) connects secondary-structure elements (e.g., β-strands J and K) within D1. This helps stabilize the orientation of D1 relative to D2, contributing to the proper positioning of the glycine-binding pocket. Its conformational stability helps define the “rigid” half of the clamshell. Loop 2 (in Domain 2) is more flexible and directly involved in the hinge motion between D1 and D2.
It plays a major role in closing the clamshell during agonist binding. Functionally, Loop 2 interacts with other receptor subunits (especially the GluN2 NTD α5 helix; Bleier et al., 2024), serving as an allosteric coupling site that tunes open probability (Po).

<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    The ligand-binding domain (LBD) of the NMDA receptor subunit NR1 (GluN1) undergoes substantial conformational rearrangements during activation, transitioning from an open “apo” state to a closed “agonist-bound” state upon glycine and glutamate binding. Bleier and colleagues (2024) directly compared NR1/NR2A-D receptor conformations in apo-like, partially agonized, and fully agonized states using single-molecule FRET. They found that in the absence of agonists (apo state), the NR1/NR2B receptor adopts a loosely packed, flexible conformation that likely contributes to its lower open probability and slower gating (Bleier et al., 2024). When fully agonized (glycine + glutamate) conformation, all subtypes converge to a similar, compact NR1-LBD arrangement (Bleier et al., 2024). However, the NR1-NTD conformation remains subtype-specific, even when fully agonized, the NR1/NR2B NTD stays relatively open, suggesting that NTD separation correlates with its lower open probability (Bleier et al., 2024).</p>

<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    Bleier and colleagues (2024) directly measured distances between NR1-LBD lobes (site D677) and found that agonist binding (glycine + glutamate) causes the LBDs to move closer together, consistent with D1–D2 cleft closure. Though the local backbone motions in the D1–D2 lobes (cleft closure) remain a universal gating step, the efficiency of mechanical coupling from that local motion to the pore varies with GluN2 subtype (Bleier et al., 2024). In NR1/NR2A receptors, ligand-induced closure of the GluN1 LBD transmits efficiently through the NTD–LBD interface to open the ion channel, resulting in a high open probability (Po ≈ 0.5), whereas in NR1/NR2B receptors, the LBDs still close upon agonist binding but the NTDs remain more splayed, weakening interdomain coupling and producing a lower open probability (Bleier et al., 2024).</p>
    <p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
   A lower open probability (Po) means the NMDA receptor channel spends less time in the open, ion-conducting state, even when both agonists (glycine and glutamate) are bound. For NR1/NR2B, this has several physiological and pharmacological implications including slower channel gating kinetics, prolonged but weaker synaptic currents, and greater susceptibility to modulation by allosteric regulators such as zinc, protons, and ifenprodil (Bleier et al., 2024). Functionally, this allows NR1/NR2B receptors to shape slower, more sustained calcium signaling that supports processes like synaptic plasticity and developmental refinement, while pharmacologically making them a prime target for subtype-selective inhibitors that exploit their unique, low-Po conformational landscape.</p>
    <p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    This hinge region represents a structural “control point” where small molecules, antibodies, or allosteric modulators could stabilize specific conformations (open, closed, or intermediate). By identifying residues whose backbone geometry changes most dramatically between states, we can pinpoint flexible or energetically sensitive sites that define the LBD’s conformational energy landscape. These hinge and loop residues, therefore, may serve as allosteric drug-binding hotspots, allowing selective modulation of NR1 activity.
        <p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    Building on these structural and conformational findings, I hypothesize that hinge and loop residues at the D1–D2 interface of the NR1 ligand-binding domain exhibit distinct φ/ψ angle clusters between apo and agonist-bound states. These backbone torsion differences are expected to reflect the clamshell-like opening and closing motion that underlies ligand-induced LBD rearrangement and mechanical coupling to the channel gate. Mapping φ/ψ transitions at the D1–D2 interface helps reveal where and how NR1 can be conformationally “tuned,” a key step in evaluating its structural and functional druggability.
</p>

    Methods:
  </h3> 
  <h4 style="color: #34495e; font-family: 'Georgia', serif; font-weight: normal; font-size: 15px; margin-top: 0;">
    <b>Sequence Used</b>: PDB 7EU8 (human NR1 LBD, chains A & C) and PDB 4KCC (rat NR1 LBD, chain A)
  </h4>

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; text-align: justify;">
    The agonist-bound conformation of the GluN1 ligand-binding domain (LBD) was extracted from chains A and C of the full-length human GluN1/GluN2B NMDA receptor (PDB 7EU8), which was solved with glycine and glutamate bound. The apo (unliganded) reference was represented by the rat GluN1 LBD (PDB 4KCC, chain A) in the open-cleft state. Human and rat NR1 share ~99% sequence identity across the LBD, particularly within the hinge and loop regions, validating 4KCC as a model for the unbound state.
  </p>

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; text-align: justify;">
    φ (phi) and ψ (psi) backbone dihedral angles were computed directly from the atomic coordinates using Biopython’s <code>PPBuilder().build_peptides()</code> and <code>get_phi_psi_list()</code> functions. Each dihedral was defined geometrically by four consecutive backbone atoms—C(i–1)–N(i)–Cα(i)–C(i) for φ and N(i)–Cα(i)–C(i)–N(i+1) for ψ—and returned in degrees within the range (–180°, +180°]. Residues lacking complete backbone atoms were excluded. For each structure, a sequence of angle-defined residues was generated, filtered for valid φ/ψ pairs, and converted into a one-letter sequence representation.
  </p>

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; text-align: justify;">
    To ensure residue-level correspondence, the angle-available sequences from 7EU8 and 4KCC were globally aligned using Biopython’s <code>PairwiseAligner</code> with the BLOSUM62 substitution matrix (gap open = –10, extend = –0.5). The resulting alignment was used to pair residues one-to-one, reconstructing gapped sequences robustly across Biopython versions. For each aligned residue pair, Δφ and Δψ were computed as circular differences wrapped into (–180°, +180°], and an average deviation metric (|Δ| mean = (|Δφ| + |Δψ|)/2) was calculated to quantify backbone rearrangement. 
  </p>

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; text-align: justify;">
    The resulting dataset (<code>NR1_phi_psi_comparison_ANGLE_ALIGNED.csv</code>) contains one row per aligned residue pair, reporting the bound and apo φ/ψ values, the angular differences, and region assignments (S1, S2, hinge, or loop). These values were used to identify localized backbone motions associated with ligand-induced clamshell closure of the NR1 LBD.
  </p>

    1. Load 7EU8 human NMDA Receptor (NR1/NR2B subtype) and 4KCC rat NR1 LBD and save each of the LBDs

```python
import pymol
pymol.finish_launching(['pymol', '-qc'])
from pymol import cmd

cmd.reinitialize()

# Fetch both structures
cmd.fetch('7EU8')  # human GluN1/2B full receptor
cmd.fetch('4KCC')  # rat NR1 LBD (apo)

# Select the LBD region for chain A of 7EU8 (approximate S1 + S2)
cmd.select('NR1_LBD_A_7EU8', '7EU8 and chain A and polymer.protein and (resi 400-545 or resi 660-800)')
cmd.save('hNR1_LBD_A_7EU8.pdb', 'NR1_LBD_A_7EU8')
print("Saved hNR1_LBD_A_7EU8.pdb")

# Save 4KCC chain A (the deposited LBD)
cmd.select('NR1_LBD_A_4KCC', '4KCC and chain A and polymer.protein')
cmd.save('rNR1_LBD_A_4KCC.pdb', 'NR1_LBD_A_4KCC')
print("Saved rNR1_LBD_A_4KCC.pdb")

cmd.quit()
```

<div style="border-left:6px solid #5dade2; border-radius:6px; padding:4px 0; margin-top:20px; margin-bottom:20px;">
  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; margin:0 0 0 20px;">
      2. Align sequences and compare the actual sequences in PDBs to ensure similarity: 7EU8 chain A (human NR1 within the full receptor) vs 4KCC chain A (rat NR1 LBD, apo).</p>
</div>

```python
from Bio import Align
from Bio.Align import substitution_matrices
from Bio.PDB import PDBParser, PPBuilder
from IPython.display import display, HTML

def get_sequence(pdb_path, chain_id):
    """Return one-letter AA sequence for given chain."""
    parser = PDBParser(QUIET=True)
    struct = parser.get_structure('x', pdb_path)[0]
    chain = struct[chain_id]
    ppb = PPBuilder()
    seqs = [str(pp.get_sequence()) for pp in ppb.build_peptides(chain, aa_only=True)]
    return "".join(seqs)

seq_7eu8 = get_sequence('hNR1_LBD_A_7EU8.pdb', 'A')
seq_4kcc = get_sequence('rNR1_LBD_A_4KCC.pdb', 'A')

print(f"7EU8 (human NR1 LBD): {len(seq_7eu8)} residues")
print(f"4KCC (rat NR1 LBD):   {len(seq_4kcc)} residues")
print("\n")

aligner = Align.PairwiseAligner()
aligner.mode = "global"
aligner.substitution_matrix = Align.substitution_matrices.load("BLOSUM62")
alignment = aligner.align(seq_7eu8, seq_4kcc)[0]
print(alignment)

# --- Get gapped strings robustly across Biopython versions ---
try:
    # Newer Biopython
    seqA, seqB = str(alignment.seqA), str(alignment.seqB)
except AttributeError:
    # Older Biopython: the alignment is sliceable
    seqA, seqB = map(str, alignment[:2])

seq_7eu8, seq_4kcc = map(str, alignment[:2])

# --- Identity stats ---
identical = sum(a == b for a, b in zip(seq_7eu8, seq_4kcc) if a != '-' and b != '-')
aligned_positions = sum(a != '-' and b != '-' for a, b in zip(seq_7eu8, seq_4kcc))
gapsA = seq_7eu8.count('-')
gapsB = seq_4kcc.count('-')
identity_pct = 100.0 * identical / aligned_positions if aligned_positions else 0.0

print("\nAlignment statistics:")
print(f"  Aligned length        : {len(seqA)}")
print(f"  Positions aligned     : {aligned_positions}")
print(f"  Identical residues    : {identical}")
print(f"  Sequence identity     : {identity_pct:.2f}%")
print(f"  Gaps in 7EU8 sequence : {gapsA}")
print(f"  Gaps in 4KCC sequence : {gapsB}")
```

  <h3 style="color:#7f8c8d; font-family:'Georgia',serif; font-weight:normal; font-size:20px;">
    Why This Matters

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    The φ/ψ angles describe local backbone conformation — the geometric “fold” at each residue.
    By comparing φ/ψ between the apo (<b>4KCC</b>) and agonist-bound (<b>7EU8</b>) NR1 structures, we can quantify
    conformational changes that underlie domain closure and hinge motion during receptor activation.

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">

  </p>

  <div style="background-color:#f4f6f6; border-radius:8px; padding:10px 20px; font-family:'Courier New',monospace; font-size:14px; color:#2c3e50;">
    θ = atan2((b₂u · (n₁ × n₂)), (n₁ · n₂))
  </div>

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    where b₀, b₁, b₂ are the bond vectors connecting the four atoms.
    The result is reported in degrees, ranging from −180° to +180°.
  </p>

</div>


```python
import math
import numpy as np
import pandas as pd
import re
# Build df_bound / df_apo from the saved LBD PDBs

PDB_BOUND = "hNR1_LBD_A_7EU8.pdb"   # human NR1 LBD (agonist-bound)
PDB_APO   = "rNR1_LBD_A_4KCC.pdb"   # rat NR1 LBD (apo)
CHAIN_ID  = "A"                      # your selections used chain A

# minimal 3-letter -> 1-letter map (fallback 'X')
three_to_one = {
    'ALA':'A','ARG':'R','ASN':'N','ASP':'D','CYS':'C','GLN':'Q','GLU':'E','GLY':'G',
    'HIS':'H','ILE':'I','LEU':'L','LYS':'K','MET':'M','PHE':'F','PRO':'P','SER':'S',
    'THR':'T','TRP':'W','TYR':'Y','VAL':'V'
}

def res_id_string(res):
    """Return 'A:508' or 'A:508A' including insertion code."""
    chain = res.get_parent().id
    hetflag, resseq, icode = res.id  # e.g., (' ', 508, 'A')
    icode = (icode or '').strip()
    return f"{chain}:{resseq}{icode}"

def one_letter(res):
    return three_to_one.get(res.get_resname().upper(), 'X')

def compute_phi_psi_table(pdb_path, chain_id="A"):
    """
    Returns a DataFrame with columns:
      aa, res_id, phi_deg, psi_deg
    Includes rows where phi/psi may be None (NaN) if undefined; we’ll filter later.
    """
    parser = PDBParser(QUIET=True)
    struct = parser.get_structure("x", pdb_path)
    # assume first model
    model = next(struct.get_models())
    chain = model[chain_id]

    rows = []
    ppb = PPBuilder()
    # build_peptides splits on breaks; iterate all fragments
    for poly in ppb.build_peptides(chain, aa_only=True):
        phi_psi_list = poly.get_phi_psi_list()
        for (phi, psi), residue in zip(phi_psi_list, poly):
            rows.append({
                "aa": one_letter(residue),
                "res_id": res_id_string(residue),
                "phi_deg": math.degrees(phi) if phi is not None else np.nan,
                "psi_deg": math.degrees(psi) if psi is not None else np.nan,
            })
    return pd.DataFrame(rows)

df_bound = compute_phi_psi_table(PDB_BOUND, CHAIN_ID)  # 7EU8 LBD A
df_apo   = compute_phi_psi_table(PDB_APO,   CHAIN_ID)  # 4KCC LBD A

print(df_bound.head(3))
print(df_apo.head(3))
print(f"Built tables: bound n={len(df_bound)} | apo n={len(df_apo)}")
```

```python
# ---------- Helpers ----------

def angle_seq(df):
    """Return (df_filtered, seq_string) for residues with non-null φ/ψ."""
    df2 = df[pd.notna(df["phi_deg"]) & pd.notna(df["psi_deg"])].reset_index(drop=True)
    seq = "".join(df2["aa"].tolist())
    return df2, seq

def circular_diff_deg(a, b):
    """Smallest signed diff (deg) in (-180, 180]; safe on NaN."""
    if pd.isna(a) or pd.isna(b):
        return np.nan
    d = (a - b + 180.0) % 360.0 - 180.0
    return 180.0 if d == -180.0 else d

def build_gapped_robust(aln, s1, s2):
    """
    Return (gapped_s1, gapped_s2) robustly across Biopython versions.
    Prefers aln.path; falls back to aln.aligned; lastly tries seqA/seqB or slicing.
    """
    # Preferred: stepwise path
    if hasattr(aln, "path") and aln.path is not None:
        g1, g2 = [], []
        path = aln.path
        for (i1, j1), (i2, j2) in zip(path[:-1], path[1:]):
            di, dj = i2 - i1, j2 - j1
            if di == 1 and dj == 1:
                g1.append(s1[i1]); g2.append(s2[j1])
            elif di == 1 and dj == 0:
                g1.append(s1[i1]); g2.append('-')
            elif di == 0 and dj == 1:
                g1.append('-');    g2.append(s2[j1])
            else:
                step = min(di, dj)
                for k in range(step):
                    g1.append(s1[i1+k]); g2.append(s2[j1+k])
                for k in range(di - step):
                    g1.append(s1[i1+step+k]); g2.append('-')
                for k in range(dj - step):
                    g1.append('-'); g2.append(s2[j1+step+k])
        return "".join(g1), "".join(g2)

    # Fallback: block coordinates
    if hasattr(aln, "aligned") and aln.aligned is not None:
        blocks1, blocks2 = aln.aligned  # lists of (start,end) for each sequence
        g1, g2 = [], []
        i = j = 0
        for (s1_start, s1_end), (s2_start, s2_end) in zip(blocks1, blocks2):
            if s1_start > i:
                for k in range(i, s1_start):
                    g1.append(s1[k]); g2.append('-')
                i = s1_start
            if s2_start > j:
                for k in range(j, s2_start):
                    g1.append('-'); g2.append(s2[k])
                j = s2_start
            n = s1_end - s1_start
            for k in range(n):
                g1.append(s1[s1_start+k]); g2.append(s2[s2_start+k])
            i = s1_end; j = s2_end
        if i < len(s1):
            for k in range(i, len(s1)):
                g1.append(s1[k]); g2.append('-')
        if j < len(s2):
            for k in range(j, len(s2)):
                g1.append('-'); g2.append(s2[k])
        return "".join(g1), "".join(g2)

    # Last-ditch conveniences
    if hasattr(aln, "seqA") and hasattr(aln, "seqB"):
        return str(aln.seqA), str(aln.seqB)
    try:
        a, b = aln[:2]
        return str(a), str(b)
    except Exception:
        pass

    raise AttributeError("Alignment lacks .path/.aligned; cannot rebuild gapped strings safely.")

# ---------- 1) Filter to angle-available residues & build sequences ----------

dfb_ang, seqA_angles = angle_seq(df_bound)  # bound (7EU8)
dfa_ang, seqB_angles = angle_seq(df_apo)    # apo (4KCC)

print(f"Angle-available lengths: bound={len(dfb_ang)} | apo={len(dfa_ang)}")

# ---------- 2) Align the angle-available sequences ----------

aligner = Align.PairwiseAligner()
aligner.mode = "global"
aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")
aligner.open_gap_score  = -10
aligner.extend_gap_score = -0.5

aln2 = aligner.align(seqA_angles, seqB_angles)[0]
gA, gB = build_gapped_robust(aln2, seqA_angles, seqB_angles)

# ---------- 3) Walk the gapped strings and assemble the comparison table ----------

iA = iB = 0
rows = []
for k, (a, b) in enumerate(zip(gA, gB)):
    if a != '-' and b != '-':
        Arow = dfb_ang.iloc[iA]
        Brow = dfa_ang.iloc[iB]

        dphi = circular_diff_deg(Arow["phi_deg"], Brow["phi_deg"])
        dpsi = circular_diff_deg(Arow["psi_deg"], Brow["psi_deg"])

        rows.append({
            "SeqPos": iA,  # index among angle-available residues (bound)
            "Residue (bound)": f'{Arow["res_id"].split(":")[1]} {Arow["aa"]}',
            "φ_bound°": round(Arow["phi_deg"], 2),
            "ψ_bound°": round(Arow["psi_deg"], 2),

            "SeqPos (apo)": iB,  # index among angle-available residues (apo)
            "Residue (apo)": f'{Brow["res_id"].split(":")[1]} {Brow["aa"]}',
            "φ_apo°": round(Brow["phi_deg"], 2),
            "ψ_apo°": round(Brow["psi_deg"], 2),

            "Δφ": round(dphi, 2),
            "Δψ": round(dpsi, 2),
            "|Δ| mean": round((abs(dphi) + abs(dpsi)) / 2.0, 2),
        })
        iA += 1
        iB += 1
    elif a != '-' and b == '-':
        iA += 1
    elif a == '-' and b != '-':
        iB += 1
    else:
        # both gaps (rare)
        pass

cmp_df = pd.DataFrame(rows)
print(f"Comparable residues (both have angles): {len(cmp_df)}")

# ---------- 4) Preview + save ----------

from IPython.display import HTML, display
display(HTML(
    "<div style='height:380px; overflow:auto; border:1px solid #bbb; "
    "border-radius:6px; font-family:ui-monospace, Menlo, monospace; font-size:12.5px;'>"
    + cmp_df.to_html(index=False)
    + "</div>"
))

cmp_df.to_csv("NR1_phi_psi_comparison_ANGLE_ALIGNED.csv", index=False)
print("Wrote NR1_phi_psi_comparison_ANGLE_ALIGNED.csv")
```

```python
import re
import numpy as np
import pandas as pd
from IPython.display import display, HTML

# ---------- helpers ----------
def _p90(s):
    """Return the 90th percentile of a Series, ignoring NaNs."""
    s = pd.to_numeric(s, errors="coerce").dropna()
    return np.percentile(s, 90) if len(s) else np.nan

def _numeric_resnum_from_label(label: str):
    """
    Extract the integer residue number from labels like '401 V' or '508A G' -> 401 / 508.
    Used for coarse region binning.
    """
    if not isinstance(label, str):
        return np.nan
    m = re.search(r"(\d+)", label)
    return int(m.group(1)) if m else np.nan

# ---------------------------------------------------------
# 1) Global φ/ψ difference summary
# ---------------------------------------------------------
# Use NaN-safe conversions to be robust to any missing values.
abs_dphi  = np.abs(pd.to_numeric(cmp_df["Δφ"], errors="coerce"))
abs_dpsi  = np.abs(pd.to_numeric(cmp_df["Δψ"], errors="coerce"))
abs_delta = pd.to_numeric(cmp_df["|Δ| mean"], errors="coerce")

global_stats = {
    "Residues analyzed": int(len(cmp_df)),
    "Mean |Δφ|":         float(np.nanmean(abs_dphi)),
    "Mean |Δψ|":         float(np.nanmean(abs_dpsi)),
    "Mean |Δ|":          float(np.nanmean(abs_delta)),
    "Median |Δ|":        float(np.nanmedian(abs_delta)),
    "90th percentile |Δ|": float(_p90(abs_delta)),
    "Max |Δ|":           float(np.nanmax(abs_delta)),
}
global_table = (
    pd.DataFrame(global_stats.items(), columns=["Metric", "Value"])
      .assign(Value=lambda d: d["Value"].round(2))
)
display(global_table)

# ---------------------------------------------------------
# 2) Regional φ/ψ difference summary
# ---------------------------------------------------------
# Ensure we have a numeric residue index (from bound labels) to assign regions.
if "resnum_bound_num" not in cmp_df.columns:
    if "Residue (bound)" not in cmp_df.columns:
        raise KeyError("cmp_df must contain 'Residue (bound)'.")
    cmp_df["resnum_bound_num"] = cmp_df["Residue (bound)"].map(_numeric_resnum_from_label)

# Define coarse windows (edit as needed). Earlier keys in PRIORITY have precedence.
REGION_WINDOWS = {
    "hinge_S1_end":   (540, 545),
    "hinge_S2_start": (660, 665),
    # Example finer windows (uncomment & add before S1/S2 in PRIORITY to override):
    # "Loop1": (508, 514),
    # "Loop2": (730, 737),
    "S1": (400, 545),
    "S2": (660, 800),
}
REGION_PRIORITY = [
    "hinge_S1_end",
    "hinge_S2_start",
    # "Loop1", "Loop2",
    "S1",
    "S2",
]

def _assign_region(n):
    """Map numeric residue index to a coarse structural region."""
    if pd.isna(n):
        return "unknown"
    for key in REGION_PRIORITY:
        lo, hi = REGION_WINDOWS[key]
        if lo <= n <= hi:
            return key
    return "other"

cmp_df["region"] = cmp_df["resnum_bound_num"].map(_assign_region)

# Regional aggregates (NaN-safe)
region_stats = (
    cmp_df.groupby("region", dropna=False)
          .agg(
              N=("|Δ| mean", "size"),
              mean_abs_dphi=("Δφ",  lambda s: np.nanmean(np.abs(pd.to_numeric(s, errors="coerce")))),
              mean_abs_dpsi=("Δψ",  lambda s: np.nanmean(np.abs(pd.to_numeric(s, errors="coerce")))),
              mean_abs_delta=("|Δ| mean", lambda s: np.nanmean(pd.to_numeric(s, errors="coerce"))),
              sd_abs_delta=("|Δ| mean", lambda s: np.nanstd(pd.to_numeric(s, errors="coerce"), ddof=1)),
              p90_abs_delta=("|Δ| mean", _p90),
              max_abs_delta=("|Δ| mean", lambda s: np.nanmax(pd.to_numeric(s, errors="coerce"))),
          )
          .round(2)
          .sort_values("mean_abs_delta", ascending=False)
)
display(region_stats)

# ---------------------------------------------------------
# 3) Regional % residues exceeding |Δ| threshold
# ---------------------------------------------------------
threshold = 40
prop_over = (abs_delta > threshold).mean() * 100.0
print(f"Residues with |Δ| > {threshold}°: {prop_over:.1f}% of all residues\n")

region_prop = (
    cmp_df.assign(significant = abs_delta > threshold)
          .groupby("region", dropna=False)["significant"]
          .mean()
          .mul(100)
          .round(1)
          .to_frame(f"% |Δ| > {threshold}°")
          .sort_values(f"% |Δ| > {threshold}°", ascending=False)
)
display(region_prop)

# ---------------------------------------------------------
# 4) Top movers (rows with |Δ| ≥ 90th percentile), scrollable preview
# ---------------------------------------------------------
top_cutoff = float(_p90(abs_delta))
top_residues = (
    cmp_df.loc[abs_delta >= top_cutoff, ["Residue (bound)","Residue (apo)","region","Δφ","Δψ","|Δ| mean"]]
          .sort_values("|Δ| mean", ascending=False)
)

display(HTML(
    "<div style='height:480px; overflow:auto; border:1px solid #bbb; "
    "border-radius:6px; font-family:ui-monospace, Menlo, monospace; font-size:12.5px;'>"
    + top_residues.to_html(index=False)
    + "</div>"
))
```

```python
# ===== NR1 dihedral comparison: plotting suite (Matplotlib only) =====
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

OUTDIR = "fig_nr1_lbd"
os.makedirs(OUTDIR, exist_ok=True)

def _savefig(name):
    path = os.path.join(OUTDIR, name)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Saved {path}")

# -----------------------------
# 1) Distribution of |Δ| mean
# -----------------------------
plt.figure(figsize=(6,4))
cmp_df["|Δ| mean"].hist(bins=40)
plt.xlabel("|Δ| mean (deg)")
plt.ylabel("Residue count")
plt.title("NR1 LBD: Distribution of mean |Δ| (apo vs bound)")
_savefig("dist_abs_delta_mean.png")

# -----------------------------------
# 2) Ramachandran: bound vs apo
# -----------------------------------
plt.figure(figsize=(5,5))
plt.scatter(cmp_df["φ_bound°"], cmp_df["ψ_bound°"], s=8, label="bound")
plt.scatter(cmp_df["φ_apo°"],   cmp_df["ψ_apo°"],   s=8, label="apo")
plt.axhline(0, lw=0.6); plt.axvline(0, lw=0.6)
plt.xlim(-180, 180); plt.ylim(-180, 180)
plt.xlabel("φ (deg)")
plt.ylabel("ψ (deg)")
plt.title("Ramachandran: NR1 LBD bound vs apo")
plt.legend()
_savefig("rama_bound_vs_apo.png")

# -----------------------------------
# 3) Δφ vs Δψ scatter (difference space)
# -----------------------------------
plt.figure(figsize=(5,5))
plt.scatter(cmp_df["Δφ"], cmp_df["Δψ"], s=10)
plt.axhline(0, lw=0.6); plt.axvline(0, lw=0.6)
plt.xlim(-180, 180); plt.ylim(-180, 180)
plt.xlabel("Δφ (deg)")
plt.ylabel("Δψ (deg)")
plt.title("Δφ vs Δψ (wrapped to ±180°)")
_savefig("delta_phi_vs_delta_psi.png")

# ---------------------------------------------------------
# 4) Highlight top-N by |Δ| mean on bound Ramachandran
# ---------------------------------------------------------
N = 20
topN = cmp_df.nlargest(N, "|Δ| mean").copy()

plt.figure(figsize=(5,5))
plt.scatter(cmp_df["φ_bound°"], cmp_df["ψ_bound°"], s=8)
plt.scatter(topN["φ_bound°"], topN["ψ_bound°"], s=28, marker="x")
for _, r in topN.iterrows():
    label = r["Residue (bound)"]
    plt.text(r["φ_bound°"], r["ψ_bound°"], label, fontsize=7)
plt.axhline(0, lw=0.6); plt.axvline(0, lw=0.6)
plt.xlim(-180, 180); plt.ylim(-180, 180)
plt.xlabel("φ_bound (deg)")
plt.ylabel("ψ_bound (deg)")
plt.title(f"Top {N} residues by mean |Δ| (bound coords)")
_savefig("rama_bound_topN_labels.png")

# ---------------------------------------------------------
# 5) Per-region boxplot of |Δ| mean (if 'region' exists)
# ---------------------------------------------------------
if "region" in cmp_df.columns:
    order = (
        cmp_df.groupby("region")["|Δ| mean"].mean()
              .sort_values(ascending=False)
              .index.tolist()
    )
    data = [cmp_df.loc[cmp_df["region"] == r, "|Δ| mean"].dropna() for r in order]
    plt.figure(figsize=(max(6, 1.5*len(order)), 4))
    plt.boxplot(data, labels=order, notch=True)
    plt.ylabel("|Δ| mean (deg)")
    plt.title("Backbone dihedral variability by region")
    _savefig("box_abs_delta_by_region.png")

# ---------------------------------------------------------
# 6) Lollipop: top-N residues by |Δ| mean
# ---------------------------------------------------------
N = 30
tops = (cmp_df[["Residue (bound)","|Δ| mean"]]
        .sort_values("|Δ| mean", ascending=False)
        .head(N)
        .copy())
tops = tops.iloc[::-1]  # plot smallest at top for readability

plt.figure(figsize=(7, max(5, 0.25*N)))
y = np.arange(len(tops))
plt.hlines(y=y, xmin=0, xmax=tops["|Δ| mean"], lw=1.5)
plt.plot(tops["|Δ| mean"], y, "o")
plt.yticks(y, tops["Residue (bound)"])
plt.xlabel("|Δ| mean (deg)")
plt.title(f"Top {N} residues by backbone change")
plt.grid(axis="x", linestyle=":", linewidth=0.6)
_savefig("lollipop_topN_abs_delta.png")

# ---------------------------------------------------------
# 7) Cumulative fraction over threshold curve
# ---------------------------------------------------------
vals = np.sort(cmp_df["|Δ| mean"].dropna().values)
x = np.linspace(0, 100, len(vals))
plt.figure(figsize=(6,4))
plt.plot(vals, x)
plt.xlabel("|Δ| mean (deg)")
plt.ylabel("Cumulative fraction (%)")
plt.title("Cumulative distribution of backbone change")
plt.grid(linestyle=":", linewidth=0.6)
# annotate a few thresholds
for t in (20, 40, 60, 80, 100):
    frac = (vals <= t).mean() * 100
    plt.scatter([t], [frac], s=25)
    plt.text(t, frac, f"{frac:.0f}%", fontsize=8, ha="left", va="bottom")
_savefig("cdf_abs_delta_mean.png")
```

  <h3 style="color:#7f8c8d; font-family:'Georgia',serif; font-weight:normal; font-size:22px; margin-top:10px;">
    Localized hinge flexibility drives activation
  </h3>

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    The φ/ψ angle comparisons reveal that glycine binding does not trigger widespread rearrangement of the NR1 ligand-binding domain (LBD). Instead, conformational change is concentrated at the S2 hinge, where the D2 lobe pivots toward the rigid D1 lobe. This localized motion supports the clamshell closure model proposed by Furukawa & Gouaux (2003), in which D2 rotation closes the glycine-binding cleft and transmits force to the transmembrane gate through the GT linker.
  </p>

  <hr style="border:none; border-top:1px solid #d6dbdf; margin:30px 0;">

  <!-- SECTION 2 -->
  <h3 style="color:#7f8c8d; font-family:'Georgia',serif; font-weight:normal; font-size:22px; margin-top:10px;">
    Structural asymmetry ensures efficient mechanical coupling
  </h3>

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    The S1 lobe remains largely static while the S2 lobe undergoes most of the torsional change. This asymmetry allows one domain to serve as a mechanical anchor and the other as the moving lever, converting small local rotations into channel gating. Such a design ensures that ligand binding produces a precise, directed motion rather than random structural fluctuations—an essential property for reliable gating in all NMDA receptor subtypes.
  </p>

  <hr style="border:none; border-top:1px solid #d6dbdf; margin:30px 0;">

  <!-- SECTION 3 -->
  <h3 style="color:#7f8c8d; font-family:'Georgia',serif; font-weight:normal; font-size:22px; margin-top:10px;">
    Energetic precision: stability with targeted motion
  </h3>

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    The distribution of φ/ψ differences indicates that most residues maintain stable geometry, with only a few exhibiting high-magnitude rotations. This pattern reflects an energy-efficient mechanism: rigidity preserves binding-site fidelity, while limited flexibility at hinge residues provides the necessary mechanical range for domain closure. The result is a finely tuned balance between structural stability and dynamic adaptability.
  </p>

  <hr style="border:none; border-top:1px solid #d6dbdf; margin:30px 0;">

  <!-- SECTION 4 -->
  <h3 style="color:#7f8c8d; font-family:'Georgia',serif; font-weight:normal; font-size:22px; margin-top:10px;">

  </h3>

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    These findings integrate closely with single-molecule FRET data from Bleier et al. (2024), who showed that NR1/NR2A receptors couple LBD closure to pore opening more efficiently than NR1/NR2B. The present analysis provides a molecular explanation for this observation: small differences in hinge flexibility or stabilization may alter how effectively local LBD motion is transmitted through the NTD–LBD interface to the channel gate, thereby influencing open probability (Po) and gating kinetics.

Because these residues experience the largest angular changes, they are probable sites of energetic sensitivity, capable of modulating the coupling efficiency between ligand binding and channel opening.
