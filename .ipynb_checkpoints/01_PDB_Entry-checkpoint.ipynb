  <h4 style="color: #7f8c8d; font-family: 'Georgia', serif; font-weight: normal; font-size: 20px; margin-top: 0;">
    GluN1/GluN2 Heterotetramer
  </h4>

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px;">
    The NMDA receptor (NMDAR) is a heterotetrameric ionotropic glutamate receptor. It’s always built from a pair of GluN1 subunits plus a pair of regulatory GluN2 and/or GluN3 subunits.<br><br>

<b>GluN2 Subtypes</b><br>
There are four GluN2 subunits, encoded by different genes (GRIN2A–D):<br>

<b>GluN2A</b> – encoded by GRIN2A<br>
Most abundant in the adult forebrain (hippocampus, cortex). Important for synaptic plasticity and fast kinetics.<br>

<b>GluN2B</b> – encoded by GRIN2B<br>
Highly expressed during development, enriched in forebrain. Slower kinetics, implicated in learning, memory, and several neuropsychiatric disorders.<br>

<b>GluN2C</b> – encoded by GRIN2C<br>
Found mainly in cerebellum, thalamus, and some midbrain regions. Contributes to unique channel properties (reduced Mg²⁺ block).<br>

<b>GluN2D</b> – encoded by GRIN2D<br>
Expressed in midbrain, diencephalon, and some interneurons. Very slow deactivation kinetics, thought to shape excitability and extrasynaptic signaling. <br>
  </p>
