    Introduction:
  </h2>
<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    The crystal structure of the NR1 ligand-binding domain (LBD) provides a detailed molecular framework for understanding how glycine binding drives NMDA receptor activation. Furukawa & Gouaux (2003) describe the structural organization of the NR1 ligand-binding domain (LBD) as composed of two discontinuous polypeptide segments: S1 (N-terminal portion) forming lobe D1 and S2 (C-terminal portion) forming lobe D2. Together, they fold into a hinged bilobed clamshell typical of ionotri=opic glutamate receptor LBDs. Upon agonist (glycine) binding, the D2 lobe rotates and moves toward D1, resulting in cleft closure that pulls on the transmembrane gating linkers, coupling ligand binding to pore opening (Furukawa & Gouaux, 2003). This cleft closure represents a key conformational change that transmits mechanical force via the S2–M3 linker to open the ion channel gate (Furukawa & Gouaux, 2003). Furukawa & Gouaux (2003) used the crystal structure PDB 1PBQ and noted that key residues forming the binding cleft include: D1 residues (Thr518, Arg523) form the rigid anchoring surface, holding glycine in place, D2 residues (Ser688, Thr690, Glu712, Asp732) form the mobile “closing” lobe, which moves toward D1 when glycine binds, and the resulting D1–D2 cleft closure is the mechanical trigger that initiates channel gating through tension on the S2–M3 linker (Figure 1). </p>

<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    The ligand-binding domain (LBD) of the NMDA receptor subunit NR1 (GluN1) undergoes substantial conformational rearrangements during activation, transitioning from an open “apo” state to a closed “agonist-bound” state upon glycine and glutamate binding. Bleier and colleagues (2024) directly compared NR1/NR2A-D receptor conformations in apo-like, partially agonized, and fully agonized states using single-molecule FRET. They found that in the absence of agonists (apo state), the NR1/NR2B receptor adopts a loosely packed, flexible conformation that likely contributes to its lower open probability and slower gating (Bleier et al., 2024). When fully agonized (glycine + glutamate) conformation, all subtypes converge to a similar, compact NR1-LBD arrangement (Bleier et al., 2024). However, the NR1-NTD conformation remains subtype-specific, even when fully agonized, the NR1/NR2B NTD stays relatively open, suggesting that NTD separation correlates with its lower open probability (Bleier et al., 2024).</p>
    
<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    Bleier and colleagues (2024) directly measured distances between NR1-LBD lobes (site D677) and found that agonist binding (glycine + glutamate) causes the LBDs to move closer together, consistent with D1–D2 cleft closure. Though the local backbone motions in the D1–D2 lobes (cleft closure) remain a universal gating step, the efficiency of mechanical coupling from that local motion to the pore varies with GluN2 subtype (Bleier et al., 2024). In NR1/NR2A receptors, ligand-induced closure of the GluN1 LBD transmits efficiently through the NTD–LBD interface to open the ion channel, resulting in a high open probability (Po ≈ 0.5), whereas in NR1/NR2B receptors, the LBDs still close upon agonist binding but the NTDs remain more splayed, weakening interdomain coupling and producing a lower open probability (Bleier et al., 2024).</p>

```python
from Bio.PDB import MMCIFParser, PDBIO

input_cif = "7eu8.cif"
output_pdb = "7EU8.pdb"

parser = MMCIFParser(QUIET=True)
structure = parser.get_structure("7EU8", input_cif)
io = PDBIO()
io.set_structure(structure)
io.save(output_pdb)

print("Converted:", input_cif, "→", output_pdb)
```

```python
# ============================================================
# NR1 S↔T "bridge" residue finder from a static PDB structure
# ------------------------------------------------------------
# What you get:
#  - nr1_bridge_results.csv   (ranked table of bridge metrics)
#  - highlight_bridges.pml    (PyMOL script to color top residues)
#
# Requirements: Biopython, networkx, numpy, pandas (pip install biopython networkx pandas)
# ============================================================

from pathlib import Path
import math
import numpy as np
import pandas as pd
import networkx as nx
from collections import Counter, defaultdict
from Bio.PDB import PDBParser, is_aa

# -----------------------------
# 0) USER INPUTS (EDIT THESE)
# -----------------------------
pdb_path   = "7EU8.pdb"             # My Protein of Interest
nr1_chain  = "A"                    # NR1 chain ID in THIS file (e.g., "A")
contact_cutoff = 4.5                # Å; try 4.0–5.0 in sensitivity analysis
max_paths_per_pair = 50             # safety cap for enumerating multiple shortest paths

# Define your sets:
#   S = hinge/interface residues in the LBD (integers: PDB 'resseq', insertion codes ignored)
#   T = gate/linker residues leading into the transmembrane gate (e.g., S2–M3 linker, pre-M1/M3)
# TIP: if you’re unsure, start with a small seed list (10–20 residues) for each and iterate.
S_resseq = []  # e.g., [507, 508, 509, ...]
T_resseq = []  # e.g., [645, 646, 647, ...]  # adjust to your structure numbering
```

```python
# -----------------------------
# 1) Parse PDB and collect residues (NR1 only)
# -----------------------------
assert Path(pdb_path).exists(), f"PDB not found: {pdb_path}"
parser = PDBParser(QUIET=True)
structure = parser.get_structure("x", pdb_path)

# Take the first model by default
model = next(structure.get_models())

# Helper to make a stable residue key: (resseq, icode) -> "123" or "123A"
def resid_key(res):
    resseq = res.id[1]
    icode  = res.id[2].strip() or ""
    return f"{resseq}{icode}"

def resseq_only(res):
    return res.id[1]

# Collect standard amino-acid residues for the specified chain
chain = model[nr1_chain]
residues = [r for r in chain.get_residues() if is_aa(r, standard=True)]
if not residues:
    raise RuntimeError(f"No standard residues found on chain {nr1_chain}. Check chain ID.")

# Map: key -> residue object
res_by_key = {resid_key(r): r for r in residues}
```

```python
# -----------------------------
# 2) Build residue-level contact graph (weighted)
#    Edge if ANY heavy-atom pair is within contact_cutoff Å
#    Weight = 1 / d_min  (smaller distance = stronger edge)
#    (We use a CA prefilter to avoid O(N^2 * atoms^2) blowups.)
# -----------------------------
# Precompute coordinates
def heavy_atom_coords(res):
    coords = []
    for a in res.get_atoms():
        if a.element != "H":   # skip hydrogens
            coords.append(a.get_coord())
    return np.array(coords, dtype=float) if coords else np.zeros((0,3))

def ca_coord(res):
    a = res["CA"] if "CA" in res else None
    return a.get_coord() if a is not None else None

res_keys = list(res_by_key.keys())
CA = {k: ca_coord(res_by_key[k]) for k in res_keys}
XYZ = {k: heavy_atom_coords(res_by_key[k]) for k in res_keys}

G = nx.Graph()
for k in res_keys:
    # node attribute 'resseq' for easy matching with S/T lists
    G.add_node(k, resseq=resseq_only(res_by_key[k]), resname=res_by_key[k].get_resname())

# Quick spatial prefilter using CA distances
def min_heavy_distance(coords1, coords2, hard_cap=8.0):
    # Fast exit if bounding boxes far apart
    if coords1.size == 0 or coords2.size == 0:
        return math.inf
    # Simple rough cap: if centroids far, skip detailed calc
    c1 = coords1.mean(axis=0)
    c2 = coords2.mean(axis=0)
    if np.linalg.norm(c1 - c2) > hard_cap:
        return math.inf
    # Full pairwise min distance (vectorized)
    dmin = math.inf
    # chunk to avoid massive memory for very large proteins; here LBD/TMD are modest
    for i in range(coords1.shape[0]):
        d = np.linalg.norm(coords2 - coords1[i], axis=1).min()
        if d < dmin:
            dmin = d
            if dmin <= contact_cutoff:
                return dmin
    return dmin

keys = res_keys
for i in range(len(keys)):
    ki = keys[i]
    cai = CA[ki]
    if cai is None: 
        continue
    for j in range(i+1, len(keys)):
        kj = keys[j]
        caj = CA[kj]
        if caj is None:
            continue
        # Fast CA prefilter (+2 Å slack)
        if np.linalg.norm(cai - caj) > (contact_cutoff + 2.0):
            continue
        dmin = min_heavy_distance(XYZ[ki], XYZ[kj], hard_cap=contact_cutoff + 6.0)
        if dmin <= contact_cutoff:
            w = max(0.05, 1.0 / max(1e-6, dmin))  # guard against div-by-zero; lower weight => stronger
            # We store 'length' for shortest-path (Dijkstra uses 'weight' as length),
            # so invert strength: length = 1 / w = dmin
            G.add_edge(ki, kj, weight=dmin, strength=w, dmin=round(float(dmin), 3))
```

```python
# -----------------------------
# 3) Define S (hinge) and T (gate) node sets from user residue numbers
# -----------------------------

def select_by_resseq(user_list):
    """
    Accepts a mix of ints (e.g., 508) and strings with optional insertion codes (e.g., '508A').
    Matches against node keys like '508', '508A', etc. Falls back to bare resseq if only ints given.
    """
    # Build fast lookup sets
    keys_by_resseq = {}
    all_keys = set(G.nodes())
    for k, d in G.nodes(data=True):
        keys_by_resseq.setdefault(d["resseq"], set()).add(k)

    selected = set()
    missing = []

    for x in user_list:
        if isinstance(x, str):
            # try exact key match first (supports insertion code)
            if x in all_keys:
                selected.add(x)
            else:
                # also try integer interpretation (e.g., '508' -> 508)
                try:
                    rx = int(x)
                    if rx in keys_by_resseq:
                        selected |= keys_by_resseq[rx]
                    else:
                        missing.append(x)
                except ValueError:
                    missing.append(x)
        else:
            # assume int
            rx = int(x)
            if rx in keys_by_resseq:
                selected |= keys_by_resseq[rx]
            else:
                missing.append(str(x))

    if missing:
        print(f" These S/T entries didn’t match any residue keys on chain {nr1_chain}: {missing}")
        print("  Tip: open nr1_chain_inventory.csv, and copy the exact 'key' values (e.g., 508A) if insertion codes exist.")

    return selected
```

```python
# -----------------------------
# 4) Path-based bridge metrics
#    (a) Path-coverage (counts how often a node lies on S→T shortest paths)
#    (b) Betweenness centrality (subset S×T)
#    (c) Removal impact: Δ in mean S→T shortest-path length when node is removed
# -----------------------------
def all_shortest_paths_capped(G, s, t, cap=max_paths_per_pair):
    # Enumerate up to 'cap' shortest paths between s and t
    # If graph disconnected, return []
    try:
        gen = nx.all_shortest_paths(G, source=s, target=t, weight="weight")
        paths = []
        for k, p in zip(range(cap), gen):
            paths.append(p)
        return paths
    except (nx.NetworkXNoPath, nx.NodeNotFound):
        return []

# (a) Path-coverage (normalized)
all_paths = []
for s in S:
    for t in T:
        ps = all_shortest_paths_capped(G, s, t, cap=max_paths_per_pair)
        all_paths.extend(ps)

inner_nodes = [n for p in all_paths for n in p[1:-1]]  # exclude endpoints
counts = Counter(inner_nodes)
total_paths = max(1, len(all_paths))
bridge_score = {n: counts.get(n, 0) / total_paths for n in G.nodes if n not in S | T}
# (b) Betweenness (subset S×T)
# Note: normalized betweenness is fine for ranking; uses edge 'weight' as path length.
bc = nx.betweenness_centrality_subset(G, sources=S, targets=T, weight="weight", normalized=True)

# (c) Removal impact: Δ mean shortest-path length
def mean_shortest_path_len_between_sets(G, S, T):
    dists = []
    for s in S:
        # single-source Dijkstra to all T (faster than per-pair)
        try:
            length = nx.single_source_dijkstra_path_length(G, s, weight="weight")
        except nx.NetworkXError:
            continue
        for t in T:
            if t in length:
                dists.append(length[t])
    if not dists:
        return math.inf
    return float(np.mean(dists))

baseline = mean_shortest_path_len_between_sets(G, S, T)

delta_len = {}
# Only test nodes that are not endpoints
test_nodes = [n for n in G.nodes if n not in S | T]
for n in test_nodes:
    H = G.copy()
    H.remove_node(n)
    new_len = mean_shortest_path_len_between_sets(H, S, T)
    # If baseline is inf (fully disconnected), delta is 0 to avoid nonsense
    if math.isinf(baseline):
        delta = 0.0
    else:
        # If removal disconnects S↔T, treat as large penalty
        delta = (new_len - baseline) if not math.isinf(new_len) else (10.0 + baseline)
    delta_len[n] = max(0.0, float(delta))
```

```python
# -----------------------------
# 5) (Optional) community bridging — participation coefficient
#     We use a simple modularity-based community detection as a proxy.
# -----------------------------
try:
    from networkx.algorithms.community import greedy_modularity_communities
    comms = list(greedy_modularity_communities(G, weight="strength"))
    # Map node -> community id
    comm_id = {}
    for cid, cset in enumerate(comms):
        for n in cset:
            comm_id[n] = cid
    # Participation coefficient: 1 - sum_c (k_i_c / k_i)^2
    part = {}
    for n in G.nodes:
        ki = sum(G[n][nbr]["strength"] for nbr in G.neighbors(n))
        if ki <= 0:
            part[n] = 0.0
            continue
        by_c = defaultdict(float)
        for nbr in G.neighbors(n):
            c = comm_id.get(nbr, -1)
            by_c[c] += G[n][nbr]["strength"]
        part[n] = 1.0 - sum((v/ki)**2 for v in by_c.values())
except Exception:
    part = {n: np.nan for n in G.nodes}
```

```python
# -----------------------------
# 6) Assemble table and rank
# -----------------------------
def zscore(series):
    s = pd.Series(series, dtype=float)
    m, sd = s.mean(), s.std(ddof=0)
    if sd == 0 or np.isnan(sd):
        return pd.Series(np.zeros(len(s)), index=s.index)
    return (s - m) / sd

df = pd.DataFrame({
    "node": list(G.nodes()),
    "resname": [G.nodes[n]["resname"] for n in G.nodes()],
    "resseq":  [G.nodes[n]["resseq"]  for n in G.nodes()],
    "bridge_path_cov": [bridge_score.get(n, 0.0) for n in G.nodes()],
    "betweenness_ST":  [bc.get(n, 0.0)           for n in G.nodes()],
    "delta_len":       [delta_len.get(n, 0.0)     for n in G.nodes()],
    "participation":   [part.get(n, np.nan)       for n in G.nodes()],
    "is_S": [n in S for n in G.nodes()],
    "is_T": [n in T for n in G.nodes()],
})

# Exclude endpoints S/T from ranking (but keep them in table for context)
mask_rank = ~(df["is_S"] | df["is_T"])
score = (
    zscore(df.set_index("node")["bridge_path_cov"]) +
    zscore(df.set_index("node")["betweenness_ST"]) +
    zscore(df.set_index("node")["delta_len"]) +
    zscore(df.set_index("node")["participation"].fillna(df["participation"].mean()))
)
df["BridgeScore_Zsum"] = score.values

df_ranked = (
    df.sort_values(by=["is_S","is_T","BridgeScore_Zsum","bridge_path_cov","betweenness_ST","delta_len"],
                   ascending=[True, True, False, False, False, False])
)

# Save results
out_csv = "nr1_bridge_results.csv"
df_ranked.to_csv(out_csv, index=False)
print(f"Wrote: {out_csv}")
```

```python
# -----------------------------
# 7) Write a PyMOL script to highlight top bridges
# -----------------------------
top_n = 12
top_nodes = df_ranked[mask_rank].head(top_n)["node"].tolist()
top_resseq = sorted({G.nodes[n]["resseq"] for n in top_nodes})
S_list = sorted({G.nodes[n]["resseq"] for n in S})
T_list = sorted({G.nodes[n]["resseq"] for n in T})

def resi_str(nums):
    return "+".join(str(x) for x in nums) if nums else "99999"  # harmless no-op if empty

pml = f"""
reinitialize
load {Path(pdb_path).name}
hide everything
as cartoon, chain {nr1_chain}
set cartoon_transparency, 0.2
set bg_color, white

# Color context
color grey70, chain {nr1_chain}

# Color S (hinge)
select hinge_S, chain {nr1_chain} and resi {resi_str(S_list)}
color marine, hinge_S
show sticks, hinge_S
set stick_radius, 0.2, hinge_S

# Color T (gate/linkers)
select gate_T, chain {nr1_chain} and resi {resi_str(T_list)}
color orange, gate_T
show sticks, gate_T
set stick_radius, 0.2, gate_T

# Highlight top bridges
select bridges_top, chain {nr1_chain} and resi {resi_str(top_resseq)}
color red, bridges_top
show spheres, bridges_top
set sphere_scale, 0.8, bridges_top
label bridges_top and name CA, "{'{'}{resn}{'}'}{'{'}resi{'}'}"

# Tidy up view
zoom chain {nr1_chain}
"""
out_pml = "highlight_bridges.pml"
with open(out_pml, "w") as f:
    f.write(pml)
print(f"Wrote: {out_pml}")
```

```python
# -----------------------------
# 8) Quick sanity prints
# -----------------------------
print("\nTop candidates (excluding endpoints):")
display(df_ranked[mask_rank].head(10)[["resname","resseq","BridgeScore_Zsum","bridge_path_cov","betweenness_ST","delta_len","participation"]])

print("\nHeads-up:")
print(" - Sensitivity-check your contact_cutoff (4.0–5.0 Å).")
print(" - Try alternative chain IDs if NR1 isn’t A in your file.")
print(" - Provide S/T as *PDB numbering* (resseq). If insertion codes matter, this script ignores them.")
print(" - The path coverage uses up to 'max_paths_per_pair' shortest paths per S–T pair to avoid blowups.")
```

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    <b>Table 1. Global Backbone Dihedral Statistics: </b> Across 263 residues in the NR1 ligand-binding domain (LBD), the average backbone deviation between the apo (4KCC) and agonist-bound (7EU8) structures was modest overall but punctuated by localized conformational changes: Mean |Δφ| = 20.8°, Mean |Δψ| = 23.3°, giving an overall Mean |Δ| = 22.0°. Most residues deviated less than 20° (Median |Δ| = 14.5°), but the upper 10% exceeded |Δ| = 43.9°. The maximum local backbone shift reached 141.9°, indicating a few residues undergo large torsional rearrangements during the transition from apo to bound. These values suggest that while most of the LBD backbone remains geometrically conserved upon ligand binding, a subset of residues experiences substantial local rotations that likely mediate hinge motion and cleft closure.
