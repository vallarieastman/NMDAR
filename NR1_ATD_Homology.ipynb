    Introduction:
  </h2>
<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    In a recent study, Tajima et al. (2022) identified and structurally characterized functional antibodies that selectively target the amino-terminal domain (ATD) of the NMDA receptor, providing critical insight into the modular architecture and subtype specificity of NMDAR extracellular domains. Using cryo-EM and crystallography, they demonstrated that an inhibitory antibody binds to the R1 lobe of the GluN2B ATD, engaging residues such as Ser31, Glu55, Asp58, His60, and Arg67 to stabilize a closed-lobe, non-active conformation of the receptor. This extracellular epitope mapping revealed how antibody binding can allosterically regulate channel gating without interacting with the ligand-binding or transmembrane regions. The finding underscores the functional importance and structural conservation of the ATD fold across NMDAR subunits, including NR1, which shares the same bilobed, clamshell-like organization used for allosteric modulation and subunit assembly. Consequently, structural homology within the NR1 ATD suggests that similar surface features could serve as regulatory or pathogenic binding interfaces, emphasizing its role as a key determinant of receptor conformation and inter-subunit communication (Tajima et al., 2022). </p>

<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    Wang et al. (2024) resolved the structural mechanisms underlying autoantibody-mediated NMDAR clustering and internalization, offering a detailed view of how pathogenic antibodies recognize the amino-terminal domain (ATD) of the GluN1 subunit. Using cryo-electron microscopy and small-angle X-ray scattering (SAXS), the study identified two patient-derived monoclonal antibodies, mAb5F6 and mAb2G7, that specifically bind to distinct epitopes on the R1 lobe of the GluN1 ATD. mAb5F6 interacts through residues such as Thr35, Arg36, Lys37, His38, Asp100, and Thr122, while mAb2G7 recognizes the α1–β2 loop region, including Gln48, Lys51, and Arg52. These epitopes are surface-exposed and unique to GluN1, explaining their subtype specificity. Structural analyses revealed that antibody binding does not acutely alter NMDAR channel gating but instead promotes crosslinking and clustering of receptor complexes (with 1:2 or 2:1 NMDAR–mAb stoichiometry), leading to endocytosis and synaptic downregulation. Functionally, prolonged antibody exposure reduced surface NMDAR density and NMDA-evoked currents in hippocampal neurons. Together, these findings define the GluN1 ATD R1 lobe as the principal antigenic site in autoimmune encephalitis and underscore the structural homology and vulnerability of the NR1 ATD to immune recognition, a feature central to understanding receptor regulation and pathology in anti-NMDAR disorders (Wang et al., 2024).</p>

Protein sequences of the NMDA receptor subunit NR1 (GluN1; gene GRIN1) were collected from representative vertebrate species, including Homo sapiens, Mus musculus, Rattus norvegicus, Gallus gallus, Xenopus laevis, and Danio rerio. Sequences were retrieved from the UniProt and NCBI Protein databases. Each sequence was trimmed to isolate the amino-terminal domain (ATD), corresponding to the N-terminal ~380 amino acids preceding the first transmembrane segment (M1). This domain encompasses the R1 and R2 lobes that form the bilobed architecture responsible for receptor assembly and allosteric regulation.<br>

The ATD sequences were aligned using MAFFT under default parameters to generate a high-quality multiple sequence alignment (MSA). Manual inspection ensured correct alignment of secondary structure elements, particularly β-strands and α-helices forming the R1/R2 interface. A per-residue conservation score was computed from the MSA.<br>

Conservation scores were projected onto the human NR1 ATD structure using the cryo-EM coordinates reported by Wang et al. (2024) (PDB ID: 87EU). Structures were visualized in PyMOL, where residues were color-coded by conservation level (dark blue = highly conserved; white = variable). For each non-human species, a homology model of the NR1 ATD was generated using MODELLER with the human NR1 ATD (PDB ID:7EU8) structure as the template. The resulting models were superposed onto the human ATD using PyMOL’s align function to assess spatial conservation of secondary structures and antibody-binding loops.<br>

```python
from Bio import AlignIO
from collections import Counter
import math
import pandas as pd

# ------------------------------------------------------------
# 1. Load your aligned sequences (MAFFT output)
# ------------------------------------------------------------
aln = AlignIO.read("/Users/val/Projects/NMDAR/NR1_Homology/orthologs/all.fasta", "fasta")

n_seqs = len(aln)
aln_len = aln.get_alignment_length()

print(f"{n_seqs} sequences aligned, length {aln_len} columns")

# ------------------------------------------------------------
# 2. Helper functions
# ------------------------------------------------------------

def col_stats(col):
    """
    col: string of residues for one alignment column across all species.
    returns:
      - major_aa: most common non-gap residue
      - pct_identity: fraction of seqs that share that major_aa (ignoring gaps in denominator)
      - entropy: Shannon entropy over non-gap residues
    """
    # remove gaps
    residues = [aa for aa in col if aa != '-']
    if len(residues) == 0:
        return None, 0.0, 0.0  # all gaps, shouldn't really happen in ATD but safe

    counts = Counter(residues)
    total = sum(counts.values())

    # majority residue and its fraction
    major_aa, major_count = counts.most_common(1)[0]
    pct_identity = major_count / total  # 1.0 == perfectly conserved

    # Shannon entropy (bits)
    probs = [c/total for c in counts.values()]
    entropy = -sum(p * math.log2(p) for p in probs)

    return major_aa, pct_identity, entropy


def build_human_index_map(aln, human_id_guess="Human"):
    """
    We need to know: which alignment column corresponds to which HUMAN residue number?
    This lets you say "column 42 is human residue 38" instead of 'alignment column 42'.

    We'll:
    - find the human row (by ID substring match)
    - walk across its aligned sequence
    - increment residue_num only when it's not a gap
    - record mapping aln_col -> human_resnum or None if gap in human
    """
    # try to find the human sequence row
    # adjust this if your FASTA headers are different
    human_idx = None
    for i, rec in enumerate(aln):
        if "Human" in rec.id or "human" in rec.id or "Hsap" in rec.id or "Homo" in rec.id:
            human_idx = i
            break
    if human_idx is None:
        # fallback: assume first sequence is human
        human_idx = 0
        print("WARNING: couldn't auto-detect human sequence, assuming row 0 is human:", aln[0].id)

    human_seq = str(aln[human_idx].seq)

    mapping = []
    human_pos_counter = 0
    for col_i, aa in enumerate(human_seq):
        if aa != '-':
            human_pos_counter += 1
            mapping.append(human_pos_counter)
        else:
            mapping.append(None)
    return mapping, aln[human_idx].id


# ------------------------------------------------------------
# 3. Compute per-column stats and map them to human residue numbers
# ------------------------------------------------------------

human_map, human_name = build_human_index_map(aln)

rows = []
for col_i in range(aln_len):
    column = aln[:, col_i]  # all residues at this column
    major_aa, pct_id, entropy = col_stats(column)

    rows.append({
        "aln_col": col_i + 1,              # 1-based alignment column index
        "human_resnum": human_map[col_i],  # NR1 residue index in human (None if gap in human)
        "major_aa": major_aa,              # most common residue in this column
        "pct_identity": pct_id,            # 1.0 = fully conserved across non-gap seqs
        "shannon_entropy": entropy         # 0.0 = perfectly conserved
    })

df_cons = pd.DataFrame(rows)

# ------------------------------------------------------------
# 4. Clean it up: keep only positions that exist in human (i.e. not gaps in human)
# ------------------------------------------------------------
df_cons_human = df_cons.dropna(subset=["human_resnum"]).copy()
df_cons_human["human_resnum"] = df_cons_human["human_resnum"].astype(int)

# also compute a simpler "conservation_score_0to9" to mimic Jalview vibe
df_cons_human["conservation_score_0to9"] = (df_cons_human["pct_identity"] * 9).round(2)

out_csv = "/Users/val/Projects/NMDAR/NR1_Homology/conservation_data/conservation_scores.csv"
df_cons_human.to_csv(out_csv, index=False)
print("wrote:", out_csv)

from IPython.display import HTML

def show_scrollable_dataframe(df, height=400):
    """Render a scrollable HTML table inside Jupyter."""
    html = df.to_html(index=False, escape=False)
    return HTML(f"""
    <div style="
        border:1px solid #ccc;
        border-radius:8px;
        padding:8px;
        width:100%;
        height:{height}px;
        overflow:auto;
        font-family: ui-monospace, Menlo, Consolas, monospace;
        font-size: 13px;">
        {html}
    </div>
    """)

show_scrollable_dataframe(df_cons_human, height=500)
```

```python
"""
This cell writes a script model_atd_multi.py that:

Reads atd_multi.ali (your alignment),

Uses both templates (8ZH7_ATD.pdb and 7EU8_ATD.pdb),

Builds 5 models and ranks them by DOPE.

"""

# ===== Prepare target + templates + alignment for MODELLER =====
# %pip install biopython requests

from pathlib import Path
import io, subprocess, shutil, requests
from Bio.PDB import PDBParser, PDBIO, PPBuilder, Select
from Bio.Align import PairwiseAligner, substitution_matrices
from Bio import SeqIO, AlignIO
from Bio.Seq import Seq

# ------------------- paths & inputs -------------------
from pathlib import Path

# Base project directory
BASE = Path("/Users/val/Projects/NMDAR/NR1_Homology/modeller")

# Define subdirectories
DATA = BASE / "data"
OUT = BASE / "out"
DATA.mkdir(parents=True, exist_ok=True)
OUT.mkdir(parents=True, exist_ok=True)

# Target (reference structure)
TARGET_PDB = Path("/Users/val/Projects/NMDAR/7EU8.pdb")   # human NR1 ATD (target)

# Template input structures
PDB_8ZH7 = DATA / "8ZH7.pdb"     # human NR1 ATD (antibody-bound)
PDB_5TQ2 = DATA / "5TQ2.pdb"     # GluN1–GluN2A ATDs (zinc-bound)

# Template outputs (cleaned, aligned, etc.)
TEMPL_8ZH7 = OUT / "8ZH7_ATD.pdb"
TEMPL_5TQ2 = OUT / "5TQ2_ATD.pdb"

# Alignment and Modeller intermediate files
FASTA_ALL = OUT / "atd_multi.fasta"
ALN_FASTA = OUT / "atd_multi_aligned.fasta"
PIR_MULTI = OUT / "atd_multi.ali"

print(f"All outputs will be written to: {OUT.resolve()}")

# ------------------- data cleaning helper -------------------
def clean_seq(s: str) -> str:
    s = str(s).upper().replace("-", "").replace("*", "")
    trans = str.maketrans({
        "U": "C",  # selenocysteine
        "O": "K",  # pyrrolysine
        "B": "N",  # Asx
        "Z": "Q",  # Glx
        "J": "L",  # Leu/Ile
    })
    s = s.translate(trans)
    allowed = set("ACDEFGHIKLMNPQRSTVWY")  # <-- only the 20 standard AAs
    return "".join(ch if ch in allowed else "A" for ch in s)  # <-- map others to A


# ------------------- small helpers -------------------
def fetch(url: str, dest: Path):
    if not dest.exists():
        r = requests.get(url, timeout=60)
        r.raise_for_status()
        dest.write_bytes(r.content)

def chain_sequence(chain):
    ppb = PPBuilder()
    seq = ""
    for pp in ppb.build_peptides(chain, aa_only=True):
        seq += str(pp.get_sequence())
    return clean_seq(seq)  # <-- clean here

def seq_from_first_chain(pdb_path: Path):
    parser = PDBParser(QUIET=True)
    s = parser.get_structure("x", str(pdb_path))
    for m in s:
        for ch in m:
            seq = chain_sequence(ch)
            if seq:
                return ch.id, seq
    raise RuntimeError(f"No protein sequence found in {pdb_path}")

def best_chain_by_identity(pdb_path: Path, target_seq: str, min_len=100):
    parser = PDBParser(QUIET=True)
    s = parser.get_structure("x", str(pdb_path))
    tgt = clean_seq(target_seq)  # <-- ensure clean
    aligner = PairwiseAligner(); aligner.mode = "global"
    try:
        aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")
    except Exception:
        pass
    aligner.open_gap_score, aligner.extend_gap_score = -10, -0.5

    best = (None, -1.0, None)
    for m in s:
        for ch in m:
            seq = chain_sequence(ch)  # already clean
            if len(seq) < min_len:
                continue
            aln = aligner.align(seq, tgt)[0]
            try:
                A, B = str(aln.seqA), str(aln.seqB)
            except AttributeError:
                A, B = map(str, aln[:2])
            nongap = sum(1 for x, y in zip(A, B) if x != '-' and y != '-')
            ident  = sum(1 for x, y in zip(A, B) if x == y and x != '-' and y != '-') / max(nongap, 1)
            if ident > best[1]:
                best = (ch.id, ident, ch)
    return best

class KeepChainProteinOnly(Select):
    def __init__(self, chain_id):
        self.chain_id = chain_id
    def accept_chain(self, chain):
        return chain.id == self.chain_id
    def accept_residue(self, residue):
        hetflag, seqnum, icode = residue.id
        return 1 if hetflag == ' ' else 0
    def accept_atom(self, atom):
        return 1

def write_chain_only(pdb_in: Path, chain_id: str, pdb_out: Path):
    parser = PDBParser(QUIET=True)
    s = parser.get_structure("x", str(pdb_in))
    io = PDBIO(); io.set_structure(s)
    io.save(str(pdb_out), KeepChainProteinOnly(chain_id))

def run_align(in_fa: Path, out_fa: Path):
    if shutil.which("mafft"):
        aln = subprocess.check_output(["mafft", "--auto", str(in_fa)], text=True)
        out_fa.write_text(aln); return "mafft"
    if shutil.which("muscle"):
        subprocess.check_call(["muscle", "-align", str(in_fa), "-output", str(out_fa)])
        return "muscle"

    # --- Biopython fallback (pairwise progressive) ---
    recs = list(SeqIO.parse(str(in_fa), "fasta"))
    if not recs:
        raise RuntimeError("No sequences found for alignment.")

    # Ensure sequences are cleaned to 20 AAs
    for r in recs:
        r.seq = Seq(clean_seq(str(r.seq)))

    aligner = PairwiseAligner()
    aligner.mode = "global"
    aligner.substitution_matrix = None   # <-- don't use BLOSUM62
    aligner.match_score = 1.0
    aligner.mismatch_score = -1.0
    aligner.open_gap_score = -10
    aligner.extend_gap_score = -0.5

    ref = str(recs[0].seq)
    names, seqs = [recs[0].id], [ref]
    for r in recs[1:]:
        A, B = ref, str(r.seq)
        aln = aligner.align(A, B)[0]
        try:
            A_g, B_g = str(aln.seqA), str(aln.seqB)
        except AttributeError:
            A_g, B_g = map(str, aln[:2])
        ref = A_g
        names.append(r.id); seqs.append(B_g)

    with out_fa.open("w") as fh:
        fh.write(f">{names[0]}\n{ref}\n")
        for n, s in zip(names[1:], seqs[1:]):
            fh.write(f">{n}\n{s}\n")
    return "biopython_fallback"

def write_pir_multi(path: Path, seq_target: str, seq_8: str, seq_5: str):
    def fmt(seq): return (clean_seq(seq) + "*")  # ensure clean and add *
    pir = []
    pir.append("C; Multi-template alignment: 8ZH7 + 5TQ2 -> target_7EU8_ATD")
    pir.append("C; Gaps are '-' ; '*' ends each sequence for MODELLER\n")
    pir.append(">P1;8ZH7_ATD");  pir.append("structureX:8ZH7_ATD: : : : : : :"); pir.append(fmt(seq_8) + "\n")
    pir.append(">P1;5TQ2_ATD");  pir.append("structureX:5TQ2_ATD: : : : : : :"); pir.append(fmt(seq_5) + "\n")
    pir.append(">P1;target_7EU8_ATD"); pir.append("sequence:target_7EU8_ATD: : : : : : :"); pir.append(fmt(seq_target) + "\n")
    path.write_text("\n".join(pir))

# ------------------- 1) Target sequence -------------------
if not TARGET_PDB.exists():
    raise FileNotFoundError(f"Missing target PDB: {TARGET_PDB}")
t_chain, t_seq = seq_from_first_chain(TARGET_PDB)
print(f"[target] chain {t_chain}, length {len(t_seq)}")

# ------------------- 2) Fetch templates -------------------
fetch("https://files.rcsb.org/download/8ZH7.pdb", PDB_8ZH7)
fetch("https://files.rcsb.org/download/5TQ2.pdb", PDB_5TQ2)

# ------------------- 3) Pick best chain vs target; write chain-only PDBs -------------------
ch8, id8, _ = best_chain_by_identity(PDB_8ZH7, t_seq, min_len=100)
ch5, id5, _ = best_chain_by_identity(PDB_5TQ2, t_seq, min_len=100)
print(f"[8ZH7] best chain {ch8}, identity = {id8:.1%}")
print(f"[5TQ2] best chain {ch5}, identity = {id5:.1%}")

write_chain_only(PDB_8ZH7, ch8, TEMPL_8ZH7)
write_chain_only(PDB_5TQ2, ch5, TEMPL_5TQ2)

# ------------------- 4) Build multi-FASTA (cleaned sequences) -------------------
def pdb_chain_seq(pdb_path: Path):
    _, s = seq_from_first_chain(pdb_path)
    return clean_seq(s)

seq_8 = pdb_chain_seq(TEMPL_8ZH7)
seq_5 = pdb_chain_seq(TEMPL_5TQ2)
t_seq_clean = clean_seq(t_seq)

records = [
    SeqIO.SeqRecord(Seq(t_seq_clean), id="target_7EU8_ATD", description=""),
    SeqIO.SeqRecord(Seq(seq_8),       id="8ZH7_ATD",        description=f"chain {ch8}"),
    SeqIO.SeqRecord(Seq(seq_5),       id="5TQ2_ATD",        description=f"chain {ch5}"),
]
SeqIO.write(records, FASTA_ALL, "fasta")
print("Wrote", FASTA_ALL)

# ------------------- 5) Align -------------------
tool = run_align(FASTA_ALL, ALN_FASTA)
print("Alignment tool:", tool, "| Wrote", ALN_FASTA)

# ------------------- 6) Write PIR for MODELLER -------------------
aln = AlignIO.read(ALN_FASTA, "fasta")
seqs = {rec.id: str(rec.seq) for rec in aln}
write_pir_multi(PIR_MULTI, seqs["target_7EU8_ATD"], seqs["8ZH7_ATD"], seqs["5TQ2_ATD"])
print("Ready for MODELLER ->", PIR_MULTI, "with templates:", TEMPL_8ZH7.name, "and", TEMPL_5TQ2.name)
```

```python
# Rebuild PIR with exact PDB sequences and explicit residue numbering / chain IDs
from pathlib import Path
from Bio.PDB import PDBParser, PPBuilder

OUT = Path("out")
TEMPL_8ZH7 = OUT / "8ZH7_ATD.pdb"
TEMPL_5TQ2 = OUT / "5TQ2_ATD.pdb"
TARGET_PDB = OUT / "7EU8_NR1_ATD_only.pdb"   # target (from 7EU8 ATD)
PIR_PATH   = OUT / "atd_multi.ali"

if not (TEMPL_8ZH7.exists() and TEMPL_5TQ2.exists() and TARGET_PDB.exists()):
    raise FileNotFoundError("Missing one of: out/8ZH7_ATD.pdb, out/5TQ2_ATD.pdb, out/7EU8_NR1_ATD_only.pdb")

parser = PDBParser(QUIET=True)
ppb = PPBuilder()

def chain_and_seq_and_range(pdb_path):
    """Return (chain_id, sequence, first_resnum, last_resnum) for the FIRST protein chain with residues."""
    s = parser.get_structure("x", str(pdb_path))
    for model in s:
        for ch in model:
            # Build the sequence directly from residues (PPBuilder gives standard one-letter codes)
            seq = ""
            for pp in ppb.build_peptides(ch, aa_only=True):
                seq += str(pp.get_sequence())
            if not seq:
                continue
            # Residue number range among standard residues (hetflag == ' ')
            nums = [res.id[1] for res in ch if res.id[0] == ' ']
            if not nums:
                continue
            return ch.id, seq, min(nums), max(nums)
    raise RuntimeError(f"No protein chain with residues found in {pdb_path}")

# Extract exact info from the PDBs MODELLER will read
ch8, seq8, start8, end8 = chain_and_seq_and_range(TEMPL_8ZH7)
ch5, seq5, start5, end5 = chain_and_seq_and_range(TEMPL_5TQ2)
cht, seqt, startt, endt = chain_and_seq_and_range(TARGET_PDB)

# Minimal global alignment just to keep them in register (prefer MAFFT/MUSCLE if available)
# We don't "clean" letters; we assume standard 20 AAs in these ATD PDBs.
from Bio.Align import PairwiseAligner
aligner = PairwiseAligner(); aligner.mode = "global"
aligner.match_score = 1; aligner.mismatch_score = -1; aligner.open_gap_score = -10; aligner.extend_gap_score = -0.5

# Align template 8ZH7 to target
aln1 = aligner.align(seqt, seq8)[0]
try:
    tgt_aln, t8_aln = str(aln1.seqA), str(aln1.seqB)
except AttributeError:
    tgt_aln, t8_aln = map(str, aln1[:2])

# Align template 5TQ2 to target (align each to the original target, not progressively, to avoid drift)
aln2 = aligner.align(seqt, seq5)[0]
try:
    _, t5_aln = str(aln2.seqA), str(aln2.seqB)
except AttributeError:
    _, t5_aln = map(str, aln2[:2])

# Ensure all three strings are same length by padding, if needed
L = max(len(tgt_aln), len(t8_aln), len(t5_aln))
def pad(s, n): return s + ("-" * (n - len(s)))
tgt_aln = pad(tgt_aln, L)
t8_aln  = pad(t8_aln,  L)
t5_aln  = pad(t5_aln,  L)

# MODELLER PIR: EXACTLY 10 fields (9 colons). Use explicit residue numbers + chain IDs.
# Format: structureX:<code>:<start>:<chain>:<end>:<chain>:<blank>:<blank>:<blank>:<blank>
hdr8 = f"structureX:8ZH7_ATD:{start8}:{ch8}:{end8}:{ch8}:::::"
hdr5 = f"structureX:5TQ2_ATD:{start5}:{ch5}:{end5}:{ch5}:::::"
hdrt = f"sequence:target_7EU8_ATD:{startt}:{cht}:{endt}:{cht}:::::"

def star(s): return s.rstrip("*") + "*"

lines = []
lines.append("C; Multi-template alignment (numbers and chains explicit)")
lines.append("C; Headers below have exactly 10 fields (9 colons).\n")

lines += [">P1;8ZH7_ATD", hdr8, star(t8_aln), ""]
lines += [">P1;5TQ2_ATD", hdr5, star(t5_aln), ""]
lines += [">P1;target_7EU8_ATD", hdrt, star(tgt_aln), ""]

PIR_PATH.write_text("\n".join(lines))
print("Wrote:", PIR_PATH)
print("8ZH7 header:", hdr8)
print("5TQ2 header:", hdr5)
print("target header:", hdrt)
print("Lengths (target, 8ZH7, 5TQ2):", len(tgt_aln), len(t8_aln), len(t5_aln))
print("First 60 target aln:", tgt_aln[:60])
print("First 60 8ZH7   aln:", t8_aln[:60])
print("First 60 5TQ2   aln:", t5_aln[:60])
```

```python
"""
This cell writes a script model_atd_multi.py that:

Reads atd_multi.ali (your alignment),

Uses both templates (8ZH7_ATD.pdb and 5TQ2_ATD.pdb),

Builds 5 models and ranks them by DOPE.
"""
# ------------------- fix PIR headers -------------------
from pathlib import Path

# Define absolute base/output path
BASE = Path("/Users/val/Projects/NMDAR/NR1_Homology/modeller")
OUT = BASE / "out"
OUT.mkdir(parents=True, exist_ok=True)

# Locate PIR file inside the modeller/out directory
pir = OUT / "atd_multi.ali"

# Read, sanitize, and rewrite the PIR alignment file
txt = pir.read_text().splitlines()

def fix_header(line: str) -> str:
    """Ensure Modeller PIR headers have exactly four trailing colons."""
    if (
        line.startswith("structureX:8ZH7_ATD:") or
        line.startswith("structureX:5TQ2_ATD:") or
        line.startswith("sequence:target_7EU8_ATD:")
    ):
        # Collapse any run of >=4 trailing colons to exactly 4
        while line.endswith(":::::"):
            line = line[:-1]  # trim one colon at a time
    return line

fixed = [fix_header(l) for l in txt]

# Write corrected alignment back to the same output directory
pir.write_text("\n".join(fixed))

# Sanity check: print header lines + colon counts
for l in fixed:
    if l.startswith(("structureX:8ZH7_ATD:", "structureX:5TQ2_ATD:", "sequence:target_7EU8_ATD:")):
        print(l, "| colons =", l.count(":"))
```

```python
"""
find all MODELLER models (target_7EU8_ATD.B*.pdb)

read DOPE (or fallback to MolPDF) directly from the PDB “REMARK 6” lines

pick the best model (lowest DOPE; else lowest MolPDF; else lowest CA-RMSD vs target)

superpose it onto your out/7EU8_NR1_ATD_only.pdb

save:

best_model.pdb

best_model_superposed_on_7EU8_ATD.pdb

view_best_model_vs_target.pml (PyMOL script)

model_scores.csv (scores for all models)
"""

# ------------------- pick best model, superpose, and write outputs -------------------
from pathlib import Path
import re
import statistics as stats
from Bio.PDB import PDBParser, Superimposer, PDBIO

# Base modeller directory
MODELLER_DIR = Path("/Users/val/Projects/NMDAR/NR1_Homology/modeller")
MODELLER_DIR.mkdir(parents=True, exist_ok=True)

# Target ATD-only PDB (in parent NR1_Homology folder)
TARGET = Path("/Users/val/Projects/NMDAR/7EU8.pdb")
if not TARGET.exists():
    raise FileNotFoundError(f"Missing target PDB: {TARGET}")

# Find MODELLER models
models = sorted(MODELLER_DIR.glob("target_7EU8_ATD.B*.pdb"))
if not models:
    raise RuntimeError("No MODELLER models found (expected target_7EU8_ATD.B9999xxxx.pdb).")

# ------------------- helpers -------------------
def parse_scores_from_pdb(pdb_path: Path):
    """Extract DOPE and MolPDF from REMARK lines."""
    d = {"path": str(pdb_path), "name": pdb_path.name, "dope": None, "molpdf": None}
    with open(pdb_path, "r", errors="ignore") as fh:
        for line in fh:
            if line.startswith("REMARK   6"):
                m = re.search(r"DOPE(?:\s+score)?:\s*([-+]?\d+(?:\.\d+)?)", line, re.I)
                if m:
                    d["dope"] = float(m.group(1))
                m = re.search(r"MolPDF[:=]\s*([-+]?\d+(?:\.\d+)?)", line, re.I)
                if m:
                    d["molpdf"] = float(m.group(1))
    return d

def load_first_chain_CA_atoms(pdb_path: Path):
    """Return structure and CA atoms of first chain."""
    p = PDBParser(QUIET=True)
    s = p.get_structure("x", str(pdb_path))
    model = next(iter(s))
    chain = next(iter(model))
    cas = [res['CA'] for res in chain if res.id[0] == ' ' and 'CA' in res]
    return s, cas

def superpose_on_ref(mobile_struct, ref_atoms, mob_atoms):
    """Superpose mobile_struct onto reference and return RMSD, n."""
    n = min(len(ref_atoms), len(mob_atoms))
    if n == 0:
        raise RuntimeError("No overlapping CA atoms to superpose.")
    sup = Superimposer()
    sup.set_atoms(ref_atoms[:n], mob_atoms[:n])
    sup.apply(mobile_struct.get_atoms())
    return sup.rms, n

# ------------------- select best model -------------------
rows = [parse_scores_from_pdb(m) for m in models]
dope_vals = [r["dope"] for r in rows if r["dope"] is not None]
molpdf_vals = [r["molpdf"] for r in rows if r["molpdf"] is not None]

best = None
if dope_vals:
    best = min(rows, key=lambda r: r["dope"] if r["dope"] is not None else 1e9)
elif molpdf_vals:
    best = min(rows, key=lambda r: r["molpdf"] if r["molpdf"] is not None else 1e9)
else:
    # fallback by RMSD vs target
    s_ref, ca_ref = load_first_chain_CA_atoms(TARGET)
    rmsd_list = []
    for r in rows:
        s_mob, ca_mob = load_first_chain_CA_atoms(Path(r["path"]))
        try:
            rms, n = superpose_on_ref(s_mob, ca_ref, ca_mob)
            rmsd_list.append((r, rms))
        except Exception:
            continue
    if not rmsd_list:
        raise RuntimeError("Could not compute RMSD fallback for any model.")
    best, _ = min(rmsd_list, key=lambda x: x[1])

# ------------------- superpose best model -------------------
pdb_best = Path(best["path"])
s_ref, ca_ref = load_first_chain_CA_atoms(TARGET)
s_mob, ca_mob = load_first_chain_CA_atoms(pdb_best)
rms, n = superpose_on_ref(s_mob, ca_ref, ca_mob)

# ------------------- save outputs -------------------
io = PDBIO()
best_out = MODELLER_DIR / "best_model.pdb"
best_sup = MODELLER_DIR / "best_model_superposed_on_7EU8_ATD.pdb"

io.set_structure(PDBParser(QUIET=True).get_structure("x", str(pdb_best)))
io.save(str(best_out))

io.set_structure(s_mob)
io.save(str(best_sup))

# PyMOL script
pml = MODELLER_DIR / "view_best_model_vs_target.pml"
pml.write_text(f"""
reinitialize
load {TARGET.as_posix()}, target
load {best_sup.as_posix()}, best
as cartoon, target or best
color marine, target
color orange, best
align best, target
zoom target or best
""")

# ------------------- summary report -------------------
print("\n========== MODELLER RUN SUMMARY ==========")
print(f" Working directory : {MODELLER_DIR}")
print(f" Target structure  : {TARGET.name}")
print(f" Models found      : {len(models)}")

if dope_vals:
    print(f" DOPE range        : {min(dope_vals):.2f} to {max(dope_vals):.2f}")
    print(f" DOPE mean (±sd)   : {stats.mean(dope_vals):.2f} ± {stats.stdev(dope_vals):.2f}")
if molpdf_vals:
    print(f" MolPDF range      : {min(molpdf_vals):.2f} to {max(molpdf_vals):.2f}")
    print(f" MolPDF mean (±sd) : {stats.mean(molpdf_vals):.2f} ± {stats.stdev(molpdf_vals):.2f}")

print(f"\n Best model        : {best['name']}")
print(f"   DOPE              : {best['dope']}")
print(f"   MolPDF            : {best['molpdf']}")
print(f"   RMSD vs target    : {rms:.3f} Å (n={n} Cα atoms)")

print("\n Output files:")
print(f"   Best model        → {best_out.name}")
print(f"   Superposed model  → {best_sup.name}")
print(f"   PyMOL script      → {pml.name}")
print("==========================================\n")
```
