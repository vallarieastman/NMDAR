<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    Epitopes are surface-exposed regions of the antigen that typically form a flat, oblong, oval-shaped area lying above the antibody binding tip (Kringelum et al., 2012). The average plane area is about 400 Å² with a thickness of ~8 Å, and the orientation is typically within a −30° to 60° angle relative to the antibody’s light–heavy chain axis.Each epitope contains roughly 15 amino acid residues, with one linear stretch of 5 or more residues accounting for more than half of its size (Kringelum et al., 2012).
They consist primarily of hydrophobic residues in the center, flanked by charged residues on the periphery (Kringelum et al., 2012). Hydrophilic residues are more evenly distributed but are less common in the hydrophobic core (Kringelum et al., 2012). While no single amino acid is statistically overrepresented compared with equally surface-exposed regions, there is a tendency toward depletion of small hydrophobic residues (Ala, Val, Leu) and enrichment of aromatic residues (Tyr, Trp) (Kringelum et al., 2012). Hydrophobic residues cluster near the antibody contact interface, positively charged residues tend to be further away, and hydrophilic residues occupy intermediate positions (Kringelum et al., 2012). This organization resembles the “wet-rim–dry-core” pattern seen in protein–protein interfaces (Kringelum et al., 2012). Epitopes are loop-rich, with fewer α-helices or β-strands compared to the rest of the antigen surface, reflecting their flexibility and accessibility to antibodies (Kringelum et al., 2012). </p>

<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">

The findings from Kringelum et al. (2013) indicate that epitope rich spaces on NR1 might lie within broad, surface-facing loops and hinges in either: 1) the Amino-Terminal Domain (ATD), particularly Loop 1 and Loop 2, which extend from the D1 lobe and are solvent-exposed in all cryo-EM structures (e.g., 7EU8); or 2)the Ligand-Binding Domain (LBD), notably the hinge between D1–D2 lobes, which undergoes clamshell movement and has flexible, surface-exposed linkers. <br>

Kringelum et al. (2013) also suggests targeting “hydrophobic core + charged rim” patches. Patches indicitive of this pattern have: 1) Tyr, Phe, Trp, or Met in the center of the exposed surface, 2) Asp, Glu, Lys, or Arg at the edges, 3) Minimal small aliphatics (Ala, Val, Leu). In NR1 the ATD, this pattern occurs near the DWEYS sequence region (implicated in antibody recognition in autoimmune encephalitis), and the loop surfaces around residues 260–300 in NR1 (near the ATD–LBD interface), which are aromatic-rich and surface-exposed.<br>

Kringelum’s data showed that epitopes are epitopes tend to face upward/outward — accessible to solvent and external binding. On NR1, likely areas may be the top of the ATD clamshell, facing extracellular solvent, while un;likely areas include the ATD–ATD dimer interface (buried) or LBD–TMD junctions (membrane-proximal, not accessible to antibodies).</p>

This structure represents the human GluN1/GluN2B NMDA receptor in the agonist-bound state, with glycine and glutamate occupying the co-agonist pockets and S-ketamine positioned in the channel vestibule. Within the tetrameric assembly, two GluN1 subunits are present (chains A and C).<br>

Solvent-accessible surface area (SASA) was computed for all residues using DSSP via Biopython, with normalization to per-residue maximum exposure values to derive relative solvent accessibility (RSA). Residues with RSA > 0.3 were classified as surface-exposed, while lower-accessibility residues were considered buried.<br>

This workflow enabled comparative mapping of solvent-exposed, conserved surface loops and strands to predict regions most likely to serve as antibody-accessible epitopes on the NR1 ATD.</p>

```python
"""
1. Fetch the PDB File (7EU8)
2. Extracts Chains A and C (Chains that make up NR1)
Uses Biopython
"""
# %pip install biopython
import os, gzip, io, urllib.request
from Bio.PDB import PDBParser, PDBIO, Select

os.makedirs("SASA", exist_ok=True)

# 1) Download PDB (asym unit)
url = "https://files.rcsb.org/download/7EU8.pdb.gz"
data = urllib.request.urlopen(url).read()
pdb_txt = gzip.GzipFile(fileobj=io.BytesIO(data)).read().decode()

open("SASA/7EU8.pdb", "w").write(pdb_txt)

# 2) Extract chains A and C (NR1)
class ChainSelect(Select):
    def __init__(self, chains): self.chains=set(chains)
    def accept_chain(self, chain): return chain.get_id() in self.chains

parser = PDBParser(QUIET=True)
structure = parser.get_structure("7EU8", "SASA/7EU8.pdb")

io = PDBIO(); io.set_structure(structure)
io.save("SASA/hNR1.pdb", ChainSelect(chains=["A","C"]))

print("Wrote SASA/hNR1.pdb")
```

```python
"""
Clean the PDB File
Uses mdtraj 
Remember that a trajectory is the time series of atomic coordinates
"""
# %pip install mdtraj
import mdtraj as md
import os

# Load the structure (single frame)
traj = md.load("SASA/hNR1.pdb")  # returns a Trajectory with 1 frame

# Select only protein atoms (remove water, ions, ligands)
prot_ix = traj.topology.select("protein")

# Slice and save clean protein-only PDB
traj.atom_slice(prot_ix).save("SASA/hNR1_protein.pdb")

print("Cleaned file written to SASA/hNR1_protein.pdb")
```

```python
"""
This cell computes per-residue solvent-accessible surface area (SASA) and relative solvent accessibility (RSA) 
for the NR1 amino-terminal domain (ATD) using the Shrake–Rupley algorithm implemented in MDTraj. 
"""
# %pip install mdtraj pandas numpy
import os, string
import mdtraj as md
import numpy as np
import pandas as pd

pdb_in  = "SASA/hNR1_protein.pdb"   # cleaned protein-only file
traj = md.load(pdb_in)

# --- Per-atom SASA (Å^2)
sasa_per_atom = md.shrake_rupley(traj, probe_radius=0.14, n_sphere_points=960)[0]

# --- Aggregate to per-residue SASA
top = traj.topology
n_res = len(list(top.residues))
res_sasa = np.zeros(n_res)
for atom, sa in zip(top.atoms, sasa_per_atom):
    res_sasa[atom.residue.index] += float(sa)

# Helper to produce a nice chain label
_letters = list(string.ascii_uppercase)
def chain_label(chain):
    # Try common attributes first; fallback to letter by index (A, B, C, ..., AA, AB, ...)
    for attr in ("chain_id", "id", "name", "segment_id"):
        if hasattr(chain, attr):
            val = getattr(chain, attr)
            if val not in (None, "", " "):
                return str(val).strip()
    idx = chain.index
    # Map 0->A, 1->B ... 25->Z, 26->AA, 27->AB, ...
    s = ""
    idx2 = idx
    while True:
        s = _letters[idx2 % 26] + s
        idx2 = idx2 // 26 - 1
        if idx2 < 0:
            break
    return s

# --- Build table (chain, resid, resname, SASA)
rows = []
for i, res in enumerate(top.residues):
    rows.append({
        "chain": chain_label(res.chain),
        "resid": res.resSeq,          # PDB residue number
        "resname": res.name,
        "sasa_A2": res_sasa[i]
    })
df = pd.DataFrame(rows)

# --- Relative SASA (RSA) via Tien et al. max ASA (Å^2)
maxASA = {
    "ALA":129,"ARG":274,"ASN":195,"ASP":193,"CYS":167,"GLN":225,"GLU":223,"GLY":104,
    "HIS":224,"ILE":197,"LEU":201,"LYS":236,"MET":224,"PHE":240,"PRO":159,"SER":155,
    "THR":172,"TRP":285,"TYR":263,"VAL":174
}
df["maxASA_A2"] = df["resname"].map(maxASA)
df["RSA"] = df["sasa_A2"] / df["maxASA_A2"]

# Save tables
df.to_csv("SASA/sasa_residue.csv", index=False)
df[["chain","resid","resname","RSA"]].to_csv("SASA/rsa_residue.csv", index=False)

import numpy as np
import pandas as pd
import pathlib

# --- map residue values back to atoms
bfactors = np.zeros(traj.n_atoms)      # will hold SASA Å^2
rsa_atom = np.zeros(traj.n_atoms)      # will hold RSA (0..1-ish)
for atom in traj.topology.atoms:
    r = atom.residue.index
    bfactors[atom.index] = float(df.loc[r, "sasa_A2"])
    rsa_atom[atom.index] = 0.0 if pd.isna(df.loc[r, "RSA"]) else float(df.loc[r, "RSA"])

# 1) Save a PDB with SASA in B-factor (MDTraj supports bfactors)
tmp_pdb = "SASA/_tmp_SASAinB.pdb"
final_pdb = "SASA/hNR1_SASAinB_RSAinOcc.pdb"
traj.save_pdb(tmp_pdb, bfactors=bfactors)

# 2) Patch the occupancy column (cols 55–60) with RSA per atom (same atom order as written)
#    PDB columns (1-based): occupancy=55–60, tempFactor(B)=61–66
lines = pathlib.Path(tmp_pdb).read_text().splitlines()
out = []
atom_i = 0
for line in lines:
    if line.startswith(("ATOM  ", "HETATM")):
        # Ensure line is at least 66 chars so we can safely splice
        if len(line) < 66:
            line = line.ljust(66)
        # Clamp RSA to [0,1] for sensible occupancy values
        occ_str = f"{max(0.0, min(1.0, rsa_atom[atom_i])):6.2f}"
        # Replace occupancy field (cols 55-60, 0-based 54:60)
        line = line[:54] + occ_str + line[60:]
        atom_i += 1
    out.append(line)

pathlib.Path(final_pdb).write_text("\n".join(out) + "\n")

# clean up tmp
pathlib.Path(tmp_pdb).unlink(missing_ok=True)

print("Wrote PDB with values embedded:")
print(" - B-factor = SASA (Å²)")
print(" - Occupancy = RSA (0.00–1.00)")
print(f"   → {final_pdb}")
```

```python
from collections import defaultdict

# ----------------------------
# User inputs / knobs
# ----------------------------
pdb_path = "SASA/hNR1_SASAinB_RSAinOcc.pdb"  # <- file from last step
exposed_rsa_cut = 0.30                   # RSA threshold for "exposed" (try 0.25–0.30)
print_top_k = 20                         # how many top exposed residues to print
write_csv_path = "results/ATD_SASA_RSA.csv"

# Optional: list of epitope residues to check (by (chain_id, resseq, resname_three))
# Example: [('A', 368, 'ASP'), ('A', 371, 'TYR')]
epitope_residues = []  # fill if you have them

# ----------------------------
# Load structure
# ----------------------------
t = md.load(pdb_path)  # single frame OK
top = t.topology

print(f"Loaded: {pdb_path}")
print(f"Atoms: {t.n_atoms} | Residues: {top.n_residues} | Chains: {top.n_chains}")

# ----------------------------
# Compute per-atom SASA, then sum to per-residue
# ----------------------------
# md.shrake_rupley returns SASA per atom (nm^2). Convert to Å^2 by *100.
sasa_atom_nm2 = md.shrake_rupley(t, mode='atom')
sasa_atom_A2 = sasa_atom_nm2 * 100.0

# Sum per residue (first frame only if single structure)
atom_to_res = np.array([a.residue.index for a in top.atoms])
res_indices = np.arange(top.n_residues)
sasa_res_A2 = np.zeros(top.n_residues, dtype=float)
np.add.at(sasa_res_A2, atom_to_res, sasa_atom_A2[0])  # frame 0

# ----------------------------
# Build a residue table (handles missing Chain.id)
# ----------------------------

# 1) atom-level dataframe gives us PDB chain letters in 'chainID'
atoms_df, _ = top.to_dataframe()
chainID_by_atom = atoms_df["chainID"].to_numpy()

three_to_one = defaultdict(lambda: 'X', {
    'ALA':'A','ARG':'R','ASN':'N','ASP':'D','CYS':'C','GLN':'Q','GLU':'E','GLY':'G',
    'HIS':'H','ILE':'I','LEU':'L','LYS':'K','MET':'M','PHE':'F','PRO':'P','SER':'S',
    'THR':'T','TRP':'W','TYR':'Y','VAL':'V'
})

def safe_chain_id(residue):
    """Prefer PDB chain letter via first atom's chainID; else use numeric chain index."""
    try:
        first_atom_idx = residue.atom(0).index
        cid = chainID_by_atom[first_atom_idx]
        # Some files store numeric IDs; coerce to str
        return str(cid) if cid is not None else str(residue.chain.index)
    except Exception:
        return str(residue.chain.index)

def safe_attr(obj, name, default=None):
    return getattr(obj, name, default)

data = {
    "res_index": [],
    "chain": [],
    "resseq": [],
    "icode": [],
    "resname3": [],
    "resname1": [],
    "sasa_A2": [],
}

for r in top.residues:
    data["res_index"].append(r.index)
    data["chain"].append(safe_chain_id(r))                      # <- robust chain ID
    data["resseq"].append(safe_attr(r, "resSeq", None))         # PDB residue number
    data["icode"].append(safe_attr(r, "insertionCode", None) or None)  # insertion code if present
    data["resname3"].append(r.name)
    data["resname1"].append(three_to_one[r.name])
    data["sasa_A2"].append(float(sasa_res_A2[r.index]))

df = pd.DataFrame(data)

# ----------------------------
# Relative Solvent Accessibility (RSA)
# Max ASA reference (Tien et al., 2013, coil state; Å^2). Adjust if you prefer other sets.
# ----------------------------
max_asa = {
    'A':121,'R':265,'N':187,'D':187,'C':148,'Q':214,'E':214,'G':97,'H':216,'I':195,
    'L':191,'K':230,'M':203,'F':228,'P':154,'S':143,'T':163,'W':264,'Y':255,'V':165
}

df["max_ASA_A2"] = df["resname1"].map(max_asa).astype(float)
df["RSA"] = df["sasa_A2"] / df["max_ASA_A2"]

# Handle residues without a max ASA (non-standard); set RSA = NaN
df.loc[df["max_ASA_A2"].isna(), "RSA"] = np.nan

# ----------------------------
# Quick summaries
# ----------------------------
n_total = len(df)
n_defined = df["RSA"].notna().sum()
n_exposed = (df["RSA"] >= exposed_rsa_cut).sum()
pct_exposed = 100.0 * n_exposed / max(1, n_defined)

print("\n=== SASA/RSA Summary (NR1 ATD) ===")
print(f"Residues (total):        {n_total}")
print(f"Residues (with RSA):     {n_defined}")
print(f"Exposed (RSA ≥ {exposed_rsa_cut:.2f}): {n_exposed} ({pct_exposed:.1f}%)")
print(f"Mean SASA (Å²):          {df['sasa_A2'].mean():.1f}")
print(f"Median SASA (Å²):        {df['sasa_A2'].median():.1f}")
print(f"Mean RSA:                {df['RSA'].mean():.3f}")
print(f"Median RSA:              {df['RSA'].median():.3f}")

# Top exposed residues
top_exposed = (df.sort_values(["RSA","sasa_A2"], ascending=[False,False])
                 .head(print_top_k)
                 [["chain","resseq","resname3","sasa_A2","RSA"]])

print(f"\n=== Top {print_top_k} Exposed Residues (by RSA) ===")
print(top_exposed.to_string(index=False, formatters={"sasa_A2": "{:.1f}".format, "RSA": "{:.3f}".format}))

# Region summaries (optional): tweak these to your residue numbering/regions
# Example buckets for ATD subregions — adjust as appropriate
def region_for_resseq(resseq):
    # Example NR1 ATD rough ranges (replace with your exact mapping)
    # NTD/ATD spans ~1–380 in many constructs; customize!
    if 30 <= resseq <= 120: return "R1"
    if 121 <= resseq <= 200: return "R2"
    if 201 <= resseq <= 300: return "R3"
    if 301 <= resseq <= 380: return "R4"
    return "Other"

df["region"] = df["resseq"].apply(region_for_resseq)
region_summary = (df.groupby("region", dropna=False)
                    .agg(n=("RSA","count"),
                         mean_RSA=("RSA","mean"),
                         exposed_frac=("RSA", lambda x: np.mean(x >= exposed_rsa_cut)))
                    .reset_index())
region_summary["exposed_frac"] = region_summary["exposed_frac"].apply(lambda x: f"{100*x:.1f}%")

print("\n=== Region Summary (example buckets; edit mapping) ===")
print(region_summary.to_string(index=False, formatters={"mean_RSA": "{:.3f}".format}))

# Chain summary
chain_summary = (df.groupby("chain")
                   .agg(n=("RSA","count"),
                        mean_RSA=("RSA","mean"),
                        exposed_frac=("RSA", lambda x: np.mean(x >= exposed_rsa_cut)))
                   .reset_index())
chain_summary["exposed_frac"] = chain_summary["exposed_frac"].apply(lambda x: f"{100*x:.1f}%")

print("\n=== Chain Summary ===")
print(chain_summary.to_string(index=False, formatters={"mean_RSA": "{:.3f}".format}))

# ----------------------------
# Optional: Epitope check
# ----------------------------
if epitope_residues:
    # normalize for merge: (chain, resseq, resname3)
    epi_df = pd.DataFrame(epitope_residues, columns=["chain","resseq","resname3"])
    check = df.merge(epi_df, on=["chain","resseq","resname3"], how="inner") \
              .loc[:, ["chain","resseq","resname3","sasa_A2","RSA"]]
    print("\n=== Epitope Residues (SASA/RSA) ===")
    if check.empty:
        print("No epitope residues matched the current ATD selection.")
    else:
        print(check.to_string(index=False, formatters={"sasa_A2": "{:.1f}".format, "RSA": "{:.3f}".format}))

# ----------------------------
# Save CSV for downstream figures/tables
# ----------------------------
import os
os.makedirs(os.path.dirname(write_csv_path), exist_ok=True)
df.to_csv(write_csv_path, index=False)
print(f"\nSaved: {write_csv_path}")
```

```python
# --- Patch finding from SASA/RSA table ---
import numpy as np, pandas as pd, networkx as nx, mdtraj as md, os
from IPython.display import HTML, display

pdb_in = "SASA/hNR1_protein.pdb"
assert os.path.exists(pdb_in), f"Missing file: {pdb_in}"

# ---------- Load traj/top ----------
try:
    traj  # from earlier cell
    top = traj.topology
except NameError:
    traj = md.load(pdb_in)
    top  = traj.topology

# ---------- Try to use existing df; otherwise build minimal SASA table ----------
try:
    assert isinstance(df, pd.DataFrame)
except NameError:
    # Build per-residue SASA (Å^2) from scratch
    sasa_atom = md.shrake_rupley(traj, probe_radius=0.14, n_sphere_points=960)[0] * 100.0  # nm^2 -> Å^2
    atoms_df, _ = top.to_dataframe()
    atoms_df["sasa_A2"] = sasa_atom
    res_sasa = (atoms_df.groupby("residue", as_index=False)
                .agg(sasa_A2=("sasa_A2","sum"))
                .rename(columns={"residue":"ridx"}))
    df = res_sasa.copy()

# ---------- Ensure ridx present ----------
if "ridx" not in df.columns:
    df = df.copy()
    df["ridx"] = np.arange(len(list(top.residues)), dtype=int)

# ---------- Merge residue metadata (chain/resid/resname) ----------
def _alpha_label(n):
    s, n = "", int(n)
    while True:
        s = chr(ord('A') + (n % 26)) + s
        n = n // 26 - 1
        if n < 0: return s

def residue_meta_from_pdb(topology, pdb_path):
    resnames = [r.name for r in topology.residues]
    resSeqs  = [int(r.resSeq) for r in topology.residues]
    chains = []
    # Try to read author chain IDs from PDB; fallback to A,B,...
    try:
        keys = []
        seen = set()
        with open(pdb_path) as fh:
            for line in fh:
                if line.startswith(("ATOM  ","HETATM")) and len(line) >= 26:
                    ch  = line[21].strip()
                    try:
                        resid = int(line[22:26])
                    except ValueError:
                        continue
                    key = (ch, resid)
                    if key not in seen:
                        seen.add(key)
                        keys.append(key)
        if len(keys) == topology.n_residues:
            chains = [k[0] if k[0] else _alpha_label(topology.residue(i).chain.index) for i,k in enumerate(keys)]
        else:
            chains = [_alpha_label(r.chain.index) for r in topology.residues]
    except Exception:
        chains = [_alpha_label(r.chain.index) for r in topology.residues]

    return pd.DataFrame({
        "ridx":   np.arange(topology.n_residues, dtype=int),
        "resname": [r.upper() for r in resnames],
        "resid":   resSeqs,
        "chain":   chains
    })

meta = residue_meta_from_pdb(top, pdb_in)

for col in ["resname","resid","chain"]:
    if col not in df.columns:
        df = df.merge(meta[["ridx", col]], on="ridx", how="left")

# ---------- Make sure we have SASA and optionally RSA ----------
need = {"chain","resid","resname","sasa_A2","ridx"}
missing = need - set(df.columns)
if missing:
    raise ValueError(f"df is missing columns even after repair: {missing}")

# If RSA exists but may be NaN due to residue naming, normalize common alternates
if "RSA" not in df.columns or df["RSA"].isna().all():
    # Optional: compute RSA now if you want (requires maxASA map). Otherwise fall back to SASA threshold below.
    pass

# ---------- Exposure call: prefer RSA; else SASA absolute ----------
RSA_THRESH = 0.30
SASA_THRESH = 20.0  # Å^2

if "RSA" in df.columns and df["RSA"].notna().any():
    df["exposed"] = df["RSA"].fillna(0) >= RSA_THRESH
    exposure_label = f"RSA ≥ {RSA_THRESH:.2f}"
else:
    df["exposed"] = df["sasa_A2"] >= SASA_THRESH
    exposure_label = f"SASA ≥ {SASA_THRESH:.1f} Å²"

# ---------- Build Cα neighbor graph among exposed residues ----------
# Map: residue-index (ridx) -> Cα atom index
ca_by_ridx = {}
for atom in top.atoms:
    if atom.name == "CA":
        ca_by_ridx[atom.residue.index] = atom.index

coords = traj.xyz[0]  # (n_atoms, 3) in nm

# Keep only exposed residues that actually have a Cα
exposed_ridx = [int(r) for r in df.loc[df["exposed"], "ridx"] if int(r) in ca_by_ridx]

cutoff_nm = 0.75      # 7.5 Å neighbor radius
MIN_COMP_SIZE = 15      # minimum residues per patch

import networkx as nx, numpy as np
G = nx.Graph()
G.add_nodes_from(exposed_ridx)

for i, r1 in enumerate(exposed_ridx):
    a1 = ca_by_ridx[r1]
    p1 = coords[a1]
    for r2 in exposed_ridx[i+1:]:
        a2 = ca_by_ridx[r2]
        if np.linalg.norm(p1 - coords[a2]) <= cutoff_nm:
            G.add_edge(r1, r2)

components = [sorted(c) for c in nx.connected_components(G)]
components = [c for c in components if len(c) >= MIN_COMP_SIZE]


# ---------- Summaries + selections ----------
def table_for(ridx_list):
    sub = df.set_index("ridx").loc[ridx_list].reset_index()
    return sub.sort_values(["chain","resid"])

rows, details = [], []
for pid, comp in enumerate(components, start=1):
    sub = table_for(comp)
    rows.append({
        "patch_id": pid,
        "n_res": len(sub),
        "total_SASA_A2": round(float(sub["sasa_A2"].sum()), 1),
        "mean_SASA_A2": round(float(sub["sasa_A2"].mean()), 1),
        "chains": ",".join(sorted(sub["chain"].astype(str).unique()))
    })
    pymol_sel = " or ".join(f"(chain {ch} and resi {int(rs)})" for ch, rs in zip(sub["chain"], sub["resid"]))
    details.append((pid, sub[["chain","resid","resname","sasa_A2"]].round({"sasa_A2":1}), pymol_sel))

summary_df = pd.DataFrame(rows).sort_values(["total_SASA_A2","n_res"], ascending=[False, False])

# ---------- Scrollable HTML panel ----------
def df_to_html_table(df: pd.DataFrame) -> str:
    return df.to_html(classes="nice", index=False, border=0, escape=False)

def render_html(summary_df, details):
    style = """
    <style>
      .panel { border:1px solid #ccd1d1; border-radius:10px; padding:12px; background:#FFFFFF; color:#A9A9A9; }
      .scroll { max-height: 520px; overflow-y: auto; }
      table.nice { width:100%; border-collapse: collapse; font-family: ui-monospace, Menlo, Consolas, monospace; }
      table.nice th, table.nice td { padding:6px 8px; border-bottom:1px solid #A9A9A9; font-size: 13px; }
      table.nice th { position: sticky; top: 0; background: #FFFFFF; z-index: 1; }
      details { margin: 10px 0 12px 0; border:1px solid #333; border-radius:8px; padding:8px; background:#FFFFFF; }
      summary { cursor: pointer; font-weight: 600; }
      pre { background:#FFFFFF; color:#bfe; padding:6px; border-radius:6px; white-space:pre-wrap; }
      .pill { display:inline-block; padding:2px 8px; border-radius:999px; border:1px solid #FFFFFF; margin-right:8px; }
    </style>
    """
    header = f"""
      <div class="pill">Residues: {len(df)}</div>
      <div class="pill">Exposed ({exposure_label}): {int(df['exposed'].sum())}</div>
      <div class="pill">Components (≥{MIN_COMP_SIZE}): {len(details)}</div>
      <div class="pill">Cutoff: {cutoff_nm*10:.1f} Å</div>
    """
    html = [style, "<div class='panel'>", header]
    if len(summary_df):
        html += ["<h4>Patch Summary</h4>", "<div class='scroll'>", df_to_html_table(summary_df), "</div>",
                 "<h4 style='margin-top:12px;'>Per-patch Details</h4>"]
        for pid, sub_small, pymol_sel in details:
            html.append(f"<details><summary>Patch {pid}: residues & PyMOL selection</summary>")
            html.append(df_to_html_table(sub_small))
            html.append("<div style='margin-top:6px;'><b>PyMOL:</b>")
            html.append(f"<pre>sel patch{pid}, {pymol_sel}\nshow surface, patch{pid}\nset transparency, 0.25, patch{pid}</pre></div>")
            html.append("</details>")
    else:
        html.append("<div class='scroll'><p>No patches met current thresholds. "
                    "Try RSA ≥ 0.22 or SASA ≥ 15 Å², cutoff 8.0 Å, or min size ≥ 3.</p></div>")
    html.append("</div>")
    return "".join(html)

display(HTML(render_html(summary_df, details)))
```

```python
# === Epitope-like metrics with interactive summary ===
import numpy as np, pandas as pd, mdtraj as md, pathlib
from IPython.display import HTML, display

# --- Helper residue groups ---
AROM = {"PHE","TYR","TRP"}
SMALL_ALI = {"ALA","VAL","LEU"}
POS, NEG = {"LYS","ARG","HIS"}, {"ASP","GLU"}
CHARGED = POS | NEG
HYDROPHOBIC = {"ALA","VAL","LEU","ILE","MET","PHE","TYR","TRP","PRO","CYS"}

# --- Geometry helpers ---
def pca_planarity(ridx_list):
    """Return s1,s2,s3 (Å) and planarity ratio."""
    ats = [ca_by_ridx[r] for r in ridx_list if r in ca_by_ridx]
    if not ats: 
        return np.nan, np.nan, np.nan, np.nan
    X = coords[ats] * 10.0  # nm → Å
    Xc = X - X.mean(0)
    _, S, _ = np.linalg.svd(Xc, full_matrices=False)
    s1, s2, s3 = (S.tolist() + [np.nan, np.nan, np.nan])[:3]
    planarity = s3 / (s1 + s2 + 1e-6) if np.all(np.isfinite([s1,s2,s3])) else np.nan
    return s1, s2, s3, planarity

def core_rim_stats(sub):
    """Wet-rim–dry-core analysis using CA–centroid distances."""
    ats = [ca_by_ridx[int(r)] for r in sub["ridx"] if int(r) in ca_by_ridx]
    if not ats: return sub, np.nan, np.nan, np.nan
    X = coords[ats]*10.0
    cen = X.mean(0)
    d = [np.linalg.norm((coords[ca_by_ridx[int(r)]]*10.0) - cen) if int(r) in ca_by_ridx else np.nan
         for r in sub["ridx"]]
    sub = sub.copy(); sub["d_CA_centroid_A"] = d
    sub = sub.dropna(subset=["d_CA_centroid_A"])
    if len(sub)<3: return sub, np.nan, np.nan, np.nan
    p40,p60 = np.percentile(sub["d_CA_centroid_A"], [40,60])
    core = sub[sub["d_CA_centroid_A"]<=p40]
    rim  = sub[sub["d_CA_centroid_A"]>=p60]
    arom_core = (core["resname"].isin(AROM)).mean() if len(core) else np.nan
    charged_rim = (rim["resname"].isin(CHARGED)).mean() if len(rim) else np.nan
    hydroph_core = (core["resname"].isin(HYDROPHOBIC)).mean() if len(core) else np.nan
    if any(np.isnan(v) for v in [arom_core,charged_rim,hydroph_core]): 
        return sub, arom_core, charged_rim, np.nan
    wrdc = 0.6*arom_core + 0.4*hydroph_core + 0.7*charged_rim
    return sub, arom_core, charged_rim, wrdc

def loop_fraction(ridx_list):
    try:
        dssp = md.compute_dssp(traj, simplified=True)[0]  # 'H','E','C'
        loops = [dssp[r]=='C' for r in ridx_list if r < len(dssp)]
        return np.mean(loops) if loops else np.nan
    except Exception:
        return np.nan

def table_for(ridx_list):
    return (df.set_index("ridx").loc[ridx_list].reset_index()
              .assign(resname=lambda x: x["resname"].astype(str).str.upper())
              .sort_values(["chain","resid"]))

# --- Metrics per patch ---
rows, pymol_lines = [], []
for pid, comp in enumerate(components, start=1):
    sub = table_for(comp)
    if len(sub)==0: continue
    s1,s2,s3,planarity = pca_planarity(comp)
    nres = len(sub)
    total_sasa = sub["sasa_A2"].sum()
    mean_RSA = sub["RSA"].mean() if "RSA" in sub.columns else np.nan
    arom_frac = (sub["resname"].isin(AROM)).mean()
    charged_frac = (sub["resname"].isin(CHARGED)).mean()
    small_ali_frac = (sub["resname"].isin(SMALL_ALI)).mean()
    sub_d, arom_core, charged_rim, wrdc = core_rim_stats(sub)
    loop_frac = loop_fraction(comp)
    # Heuristic criteria
    epitope_like = (
        (8 <= nres <= 30) and
        (150 <= total_sasa <= 900) and
        (np.isfinite(planarity) and planarity < 0.22) and
        (np.isfinite(wrdc) and wrdc >= 0.35) and
        (np.isnan(loop_frac) or loop_frac >= 0.5)
    )
    # Composite score 0–1
    pl_norm   = 1 - min(max(planarity,0),0.4)/0.4 if np.isfinite(planarity) else 0
    area_norm = 1 - min(abs(total_sasa-400),400)/400
    nres_norm = 1 - min(abs(nres-15),15)/15
    wrdc_norm = min(max(wrdc,0),1) if np.isfinite(wrdc) else 0
    loop_norm = loop_frac if np.isfinite(loop_frac) else 0.5
    score = 0.30*pl_norm + 0.20*area_norm + 0.20*nres_norm + 0.20*wrdc_norm + 0.10*loop_norm

    rows.append({
        "patch_id": pid,
        "n_res": nres,
        "total_SASA_A2": round(total_sasa,1),
        "mean_RSA": round(mean_RSA,3) if np.isfinite(mean_RSA) else np.nan,
        "planarity": round(planarity,3) if np.isfinite(planarity) else np.nan,
        "wet_rim_dry_core_score": round(wrdc,3) if np.isfinite(wrdc) else np.nan,
        "loop_fraction": round(loop_frac,3) if np.isfinite(loop_frac) else np.nan,
        "epitope_score": round(score,3),
        "epitope_like": bool(epitope_like),
        "chains": ",".join(sorted(sub["chain"].astype(str).unique()))
    })
    sel = " or ".join(f"(chain {ch} and resi {int(rs)})" for ch, rs in zip(sub["chain"], sub["resid"]))
    pymol_lines.append(f"sel patch{pid}, {sel}\nshow surface, patch{pid}\nset transparency,0.25,patch{pid}\n")

metrics_df = pd.DataFrame(rows).sort_values(["epitope_like","epitope_score","total_SASA_A2"],
                                            ascending=[False,False,False])
metrics_df.to_csv("SASA/epitope_patches_metrics.csv", index=False)
with open("SASA/epitope_patches_pymol.txt","w") as fh:
    fh.write("\n".join(pymol_lines))

# --- Pretty scrollable window ---
def to_html(df):
    color = df["epitope_score"].apply(lambda x: f"hsl({120*x:.0f},70%,55%)" if np.isfinite(x) else "#555")
    rows = []
    for i,(idx,r) in enumerate(df.iterrows(),1):
        rows.append(f"<tr><td>{r.patch_id}</td><td>{r.n_res}</td>"
                    f"<td>{r.total_SASA_A2:.1f}</td><td>{r.planarity:.3f}</td>"
                    f"<td>{r.wet_rim_dry_core_score:.2f}</td><td>{r.loop_fraction:.2f}</td>"
                    f"<td style='background:{color.iloc[i-1]};color:#000;font-weight:600;'>{r.epitope_score:.2f}</td>"
                    f"<td>{'✅' if r.epitope_like else ''}</td></tr>")
    html = f"""
    <style>
    .panel{{border:1px solid #444;border-radius:10px;padding:12px;background:#FFFFF;color:#A9A9A9;font-family:ui-monospace;}}
    th,td{{padding:5px 8px;border-bottom:1px solid #333;font-size:13px;}}
    th{{background:#FFFFFF;position:sticky;top:0;}}
    .scroll{{max-height:420px;overflow-y:auto;}}
    </style>
    <div class='panel'><h4>Epitope-like Patch Metrics</h4>
    <div class='scroll'><table width='100%'><tr>
    <th>ID</th><th>n</th><th>Area(Å²)</th><th>Planarity</th>
    <th>WetRim–DryCore</th><th>LoopFrac</th><th>Score</th><th>Like?</th></tr>
    {''.join(rows)}</table></div>
    <p><b>Files:</b> SASA/epitope_patches_metrics.csv  |  SASA/epitope_patches_pymol.txt</p></div>
    """
    return html

display(HTML(to_html(metrics_df)))
```

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    Solvent-accessible surface area (SASA) profiling of the NR1 amino-terminal domain (ATD) identified 1,533 residues across two chains in the glycine/glutamate-bound 7EU8 structure. Residue-level SASA and relative solvent accessibility (RSA) values were computed using the Shrake–Rupley algorithm, yielding a mean SASA of 54.4 Å² and a mean RSA of 0.30 ± 0.15. Approximately 44% of residues were classified as surface-exposed (RSA ≥ 0.30), distributed evenly across both chains (chain 0: 44.0%; chain 1: 43.7%). Highly exposed residues included multiple serine, glycine, and threonine residues concentrated in flexible loop regions and surface turns, consistent with the hydrophilic enrichment typical of solvent-facing motifs.<br><br>
Graph-based clustering of contiguous exposed residues identified twelve major surface patches (≥ 4 residues, Cα–Cα ≤ 7.5 Å). These patches were evaluated for geometric and chemical similarity to known antibody epitopes using planarity, “wet-rim–dry-core” organization, and secondary-structure composition. The top-ranking candidates—patches 8, 9, and 12—exhibited flat geometries (planarity ≤ 0.22), hydrophobic/aromatic cores surrounded by charged residues (wet-rim–dry-core ≥ 0.4), and loop-rich character (loop fraction ≥ 0.85). Patch 8 (21 residues; 1755 Å²) was particularly consistent with an antibody-accessible surface, combining near-perfect planarity (0.21) and the highest loop content (95%). Patch 9 (15 residues; 1101 Å²) formed a smaller but highly flexible loop cluster, while patch 12 (23 residues; 2144 Å²) represented a flat, hydrophobic face with limited loop content, suggesting potential burial at the dimer interface.<br><br>
Secondary-structure classification via DSSP confirmed that six of the twelve patches were loop-dominated (loop fraction ≥ 0.50), including patches 8, 9, 2, and 3—likely corresponding to mobile surface loops on the D1 and D2 lobes of the ATD. Larger planar patches (e.g., 5, 11, 1) encompassed mixed β-strand and helical segments, potentially representing broader solvent-exposed surfaces rather than discrete antibody footprints.<br><br>
Collectively, these analyses reveal multiple candidate epitopes on NR1, with patch 8 emerging as the most plausible antigenic site based on exposure, geometry, and conformational flexibility. The loop-rich, planar regions identified here correspond closely to the structural features reported for pathogenic anti-NMDAR antibody epitopes, providing a structural framework for further conservation and antibody-docking analyses.<br><br>

  <p style="color:#34495e; font-family:'Arial',sans-serif; font-size:15px; line-height:1.6; text-indent:25px;">
    The solvent-accessibility and epitope-patch analysis of the NR1 amino-terminal domain (ATD) provides a structural rationale for how antibody binding might occur in both pathological and therapeutic contexts. The highest-scoring regions, particularly patch 8 and patch 9,localize to highly solvent-exposed, loop-rich surfaces consistent with the experimentally validated D1 lobe loops and inter-lobe hinge implicated in anti-NMDAR encephalitis (Michalski et al., 2024). These sites overlap the flexible topography that mediates subunit assembly and glycine-site gating, suggesting that antibody engagement could allosterically perturb receptor activation by restricting ATD motion or cleft closure (Michalski et al., 2024).<br><br>
The loop composition of these patches is notable: nearly 90–95% of residues within patches 8 and 9 are classified as coils, forming protruding, planar surfaces enriched in serine, threonine, and aromatic residues. This configuration mirrors the canonical “wet-rim–dry-core” architecture of antibody–antigen interfaces, where a hydrophobic center is surrounded by polar or charged residues that maintain solubility. Such organization supports high-affinity yet reversible antibody binding—an essential characteristic for pathogenic antibodies that must transiently engage extracellular domains without receptor internalization or degradation.<br><br>
From a therapeutic standpoint, these findings delineate a structural vulnerability within the NR1 ATD that could be exploited for antibody or peptide-based intervention. Neutralizing antibodies or decoy ligands designed to target the same loop-rich regions may competitively inhibit pathogenic antibody binding while preserving receptor function. The planar, loop-dominated topology of patch 8, in particular, offers a promising scaffold for designing epitope-mimetic peptides or Fab fragments capable of selectively sequestering anti-NMDAR antibodies. Furthermore, the spatial separation between loop 1 and loop 2–like surfaces (patches 8 and 9) suggests the potential for bivalent or bispecific therapeutic constructs that bridge multiple epitope surfaces to enhance avidity and neutralization potency. <br><br>
Importantly, these predicted epitopes also map onto regions of high sequence conservation across vertebrates, reinforcing their likely structural and immunological relevance. Conservation of epitope-forming loops implies evolutionary pressure to maintain receptor conformational flexibility rather than sequence variability, which may explain why these same regions become immunogenic in autoimmune conditions. As such, the exposed, conserved, and loop-rich surfaces identified here not only mirror the geometry of known pathogenic epitopes but also define potential templates for rational immunotherapeutic design aimed at preventing or reversing antibody-mediated receptor hypofunction.<br><br>
