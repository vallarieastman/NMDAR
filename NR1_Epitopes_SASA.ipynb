<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">
    Epitopes are surface-exposed regions of the antigen that typically form a flat, oblong, oval-shaped area lying above the antibody binding tip (Kringelum et al., 2012). The average plane area is about 400 Å² with a thickness of ~8 Å, and the orientation is typically within a −30° to 60° angle relative to the antibody’s light–heavy chain axis.Each epitope contains roughly 15 amino acid residues, with one linear stretch of 5 or more residues accounting for more than half of its size (Kringelum et al., 2012).
They consist primarily of hydrophobic residues in the center, flanked by charged residues on the periphery (Kringelum et al., 2012). Hydrophilic residues are more evenly distributed but are less common in the hydrophobic core (Kringelum et al., 2012). While no single amino acid is statistically overrepresented compared with equally surface-exposed regions, there is a tendency toward depletion of small hydrophobic residues (Ala, Val, Leu) and enrichment of aromatic residues (Tyr, Trp) (Kringelum et al., 2012). Hydrophobic residues cluster near the antibody contact interface, positively charged residues tend to be further away, and hydrophilic residues occupy intermediate positions (Kringelum et al., 2012). This organization resembles the “wet-rim–dry-core” pattern seen in protein–protein interfaces (Kringelum et al., 2012). Epitopes are loop-rich, with fewer α-helices or β-strands compared to the rest of the antigen surface, reflecting their flexibility and accessibility to antibodies (Kringelum et al., 2012). </p>

<p style="text-indent:30px; color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px; line-height:1.6;">

The findings from Kringelum et al. (2013) indicate that epitope rich spaces on NR1 might lie within broad, surface-facing loops and hinges in either: 1) the Amino-Terminal Domain (ATD), particularly Loop 1 and Loop 2, which extend from the D1 lobe and are solvent-exposed in all cryo-EM structures (e.g., 7EU8); or 2)the Ligand-Binding Domain (LBD), notably the hinge between D1–D2 lobes, which undergoes clamshell movement and has flexible, surface-exposed linkers. <br>

Kringelum et al. (2013) also suggests targeting “hydrophobic core + charged rim” patches. Patches indicitive of this pattern have: 1) Tyr, Phe, Trp, or Met in the center of the exposed surface, 2) Asp, Glu, Lys, or Arg at the edges, 3) Minimal small aliphatics (Ala, Val, Leu). In NR1 the ATD, this pattern occurs near the DWEYS sequence region (implicated in antibody recognition in autoimmune encephalitis), and the loop surfaces around residues 260–300 in NR1 (near the ATD–LBD interface), which are aromatic-rich and surface-exposed.<br>

Kringelum’s data showed that epitopes are epitopes tend to face upward/outward — accessible to solvent and external binding. On NR1, likely areas may be the top of the ATD clamshell, facing extracellular solvent, while un;likely areas include the ATD–ATD dimer interface (buried) or LBD–TMD junctions (membrane-proximal, not accessible to antibodies).</p>

    PDB 7EU8 was used as the structural template for solvent-accessibility analysis. This structure represents the human GluN1/GluN2B NMDA receptor in the agonist-bound state, with glycine and glutamate occupying the co-agonist pockets and S-ketamine positioned in the channel vestibule. Within the tetrameric assembly, two GluN1 subunits are present (chains A and C).<br>

Solvent-accessible surface area (SASA) was computed for all residues using DSSP via Biopython, with normalization to per-residue maximum exposure values to derive relative solvent accessibility (RSA). Residues with RSA > 0.3 were classified as surface-exposed, while lower-accessibility residues were considered buried.<br>

This workflow enabled comparative mapping of solvent-exposed, conserved surface loops and strands to predict regions most likely to serve as antibody-accessible epitopes on the NR1 ATD.</p>

```python
"""
1. Fetch the PDB File (7EU8)
2. Extracts Chains A and C (Chains that make up NR1)
Uses Biopython
"""
# %pip install biopython
import os, gzip, io, urllib.request
from Bio.PDB import PDBParser, PDBIO, Select

os.makedirs("SASA", exist_ok=True)

# 1) Download PDB (asym unit)
url = "https://files.rcsb.org/download/7EU8.pdb.gz"
data = urllib.request.urlopen(url).read()
pdb_txt = gzip.GzipFile(fileobj=io.BytesIO(data)).read().decode()

open("SASA/7EU8.pdb", "w").write(pdb_txt)

# 2) Extract chains A and C (GluN1)
class ChainSelect(Select):
    def __init__(self, chains): self.chains=set(chains)
    def accept_chain(self, chain): return chain.get_id() in self.chains

parser = PDBParser(QUIET=True)
structure = parser.get_structure("7EU8", "SASA/7EU8.pdb")

io = PDBIO(); io.set_structure(structure)
io.save("SASA/hNR1.pdb", ChainSelect(chains=["A","C"]))

print("Wrote SASA/hNR1.pdb")
```

```python
"""
Clean the PDB File
Uses mdtraj 
Remember that a trajectory is the time series of atomic coordinates
"""
# %pip install mdtraj
import mdtraj as md
import os

# Load the structure (single frame)
traj = md.load("SASA/hNR1.pdb")  # returns a Trajectory with 1 frame

# Select only protein atoms (remove water, ions, ligands)
prot_ix = traj.topology.select("protein")

# Slice and save clean protein-only PDB
traj.atom_slice(prot_ix).save("SASA/hNR1_protein.pdb")

print("Cleaned file written to SASA/hNR1_protein.pdb")
```

```python
"""
This cell computes per-residue solvent-accessible surface area (SASA) and relative solvent accessibility (RSA) 
for the NR1 amino-terminal domain (ATD) using the Shrake–Rupley algorithm implemented in MDTraj. 
"""
# %pip install mdtraj pandas numpy
import os, string
import mdtraj as md
import numpy as np
import pandas as pd

pdb_in  = "SASA/hNR1_protein.pdb"   # cleaned protein-only file
traj = md.load(pdb_in)

# --- Per-atom SASA (Å^2)
sasa_per_atom = md.shrake_rupley(traj, probe_radius=0.14, n_sphere_points=960)[0]

# --- Aggregate to per-residue SASA
top = traj.topology
n_res = len(list(top.residues))
res_sasa = np.zeros(n_res)
for atom, sa in zip(top.atoms, sasa_per_atom):
    res_sasa[atom.residue.index] += float(sa)

# Helper to produce a nice chain label
_letters = list(string.ascii_uppercase)
def chain_label(chain):
    # Try common attributes first; fallback to letter by index (A, B, C, ..., AA, AB, ...)
    for attr in ("chain_id", "id", "name", "segment_id"):
        if hasattr(chain, attr):
            val = getattr(chain, attr)
            if val not in (None, "", " "):
                return str(val).strip()
    idx = chain.index
    # Map 0->A, 1->B ... 25->Z, 26->AA, 27->AB, ...
    s = ""
    idx2 = idx
    while True:
        s = _letters[idx2 % 26] + s
        idx2 = idx2 // 26 - 1
        if idx2 < 0:
            break
    return s

# --- Build table (chain, resid, resname, SASA)
rows = []
for i, res in enumerate(top.residues):
    rows.append({
        "chain": chain_label(res.chain),
        "resid": res.resSeq,          # PDB residue number
        "resname": res.name,
        "sasa_A2": res_sasa[i]
    })
df = pd.DataFrame(rows)

# --- Relative SASA (RSA) via Tien et al. max ASA (Å^2)
maxASA = {
    "ALA":129,"ARG":274,"ASN":195,"ASP":193,"CYS":167,"GLN":225,"GLU":223,"GLY":104,
    "HIS":224,"ILE":197,"LEU":201,"LYS":236,"MET":224,"PHE":240,"PRO":159,"SER":155,
    "THR":172,"TRP":285,"TYR":263,"VAL":174
}
df["maxASA_A2"] = df["resname"].map(maxASA)
df["RSA"] = df["sasa_A2"] / df["maxASA_A2"]

# Save tables
df.to_csv("SASA/sasa_residue.csv", index=False)
df[["chain","resid","resname","RSA"]].to_csv("SASA/rsa_residue.csv", index=False)

import numpy as np
import pandas as pd
import pathlib

# --- map residue values back to atoms
bfactors = np.zeros(traj.n_atoms)      # will hold SASA Å^2
rsa_atom = np.zeros(traj.n_atoms)      # will hold RSA (0..1-ish)
for atom in traj.topology.atoms:
    r = atom.residue.index
    bfactors[atom.index] = float(df.loc[r, "sasa_A2"])
    rsa_atom[atom.index] = 0.0 if pd.isna(df.loc[r, "RSA"]) else float(df.loc[r, "RSA"])

# 1) Save a PDB with SASA in B-factor (MDTraj supports bfactors)
tmp_pdb = "SASA/_tmp_SASAinB.pdb"
final_pdb = "SASA/hNR1_SASAinB_RSAinOcc.pdb"
traj.save_pdb(tmp_pdb, bfactors=bfactors)

# 2) Patch the occupancy column (cols 55–60) with RSA per atom (same atom order as written)
#    PDB columns (1-based): occupancy=55–60, tempFactor(B)=61–66
lines = pathlib.Path(tmp_pdb).read_text().splitlines()
out = []
atom_i = 0
for line in lines:
    if line.startswith(("ATOM  ", "HETATM")):
        # Ensure line is at least 66 chars so we can safely splice
        if len(line) < 66:
            line = line.ljust(66)
        # Clamp RSA to [0,1] for sensible occupancy values
        occ_str = f"{max(0.0, min(1.0, rsa_atom[atom_i])):6.2f}"
        # Replace occupancy field (cols 55-60, 0-based 54:60)
        line = line[:54] + occ_str + line[60:]
        atom_i += 1
    out.append(line)

pathlib.Path(final_pdb).write_text("\n".join(out) + "\n")

# (Optional) clean up tmp
pathlib.Path(tmp_pdb).unlink(missing_ok=True)

print("Wrote PDB with values embedded:")
print(" - B-factor = SASA (Å²)")
print(" - Occupancy = RSA (0.00–1.00)")
print(f"   → {final_pdb}")
```

```python
# %pip install mdtraj pandas numpy networkx
import os, mdtraj as md, numpy as np, pandas as pd, networkx as nx

os.makedirs("SASA", exist_ok=True)

pdb_in = "SASA/hNR1_protein.pdb"
rsa_df = pd.read_csv("SASA/rsa_residue.csv")       # chain,resid,resname,RSA
sasa_df = pd.read_csv("SASA/sasa_residue.csv")     # chain,resid,resname,sasa_A2  (ensure this has 'chain'!)

# ---- 1) Merge with chain + resid (IMPORTANT)
tbl = (rsa_df.merge(sasa_df[["chain","resid","sasa_A2"]], on=["chain","resid"], how="inner")
             .assign(exposed=lambda x: x["RSA"] >= 0.30))   # tweak 0.25–0.35 as needed

print(f"Residues total: {len(tbl)} | Exposed (RSA≥0.30): {int(tbl['exposed'].sum())}")

# ---- 2) Structure + Cα coordinates
traj = md.load(pdb_in)
top = traj.topology
coords = traj.xyz[0]   # nm

# map ridx -> CA atom index
ca_idx = [a.index for a in top.atoms if a.name == "CA"]
ca_by_ridx = {top.atom(i).residue.index: i for i in ca_idx}

# helper: map (chain label, resid) to mdtraj residue index
def build_label_map(topology, df_labels):
    # Try to match by exact chain label in df; fallback if necessary
    chain_label_map = {}
    # Assign the first seen df label for each mdtraj chain index
    for r in topology.residues:
        candidates = df_labels.loc[df_labels["resid"] == r.resSeq, "chain"].unique()
        if len(candidates):
            chain_label_map.setdefault(r.chain.index, str(candidates[0]))
    # fallback to numeric string if none
    for r in topology.residues:
        chain_label_map.setdefault(r.chain.index, str(r.chain.index))
    return chain_label_map

chain_label_map = build_label_map(top, tbl[["chain","resid"]])

def find_ridx(chain_label, resid_num):
    matches = [r for r in top.residues
               if chain_label_map.get(r.chain.index, "") == str(chain_label)
               and r.resSeq == int(resid_num)]
    return matches[0].index if matches else None

tbl["ridx"] = [find_ridx(ch, rs) for ch, rs in zip(tbl["chain"], tbl["resid"])]
tbl = tbl.dropna(subset=["ridx"]).astype({"ridx": int})

# ---- 3) Graph of exposed residues (edges: CA–CA <= cutoff)
exposed_ridx = set(tbl.loc[tbl["exposed"], "ridx"])
G = nx.Graph()
G.add_nodes_from(sorted(exposed_ridx))
cutoff_nm = 0.65  # 6.5 Å
nodes = sorted(exposed_ridx)

for i, r1 in enumerate(nodes):
    a1 = ca_by_ridx.get(r1, None)
    if a1 is None: 
        continue
    xyz1 = coords[a1]
    for r2 in nodes[i+1:]:
        a2 = ca_by_ridx.get(r2, None)
        if a2 is None:
            continue
        if np.linalg.norm(xyz1 - coords[a2]) <= cutoff_nm:
            G.add_edge(r1, r2)

components = [sorted(c) for c in nx.connected_components(G)]
print(f"Connected components among exposed residues: {len(components)}")

# keep components with at least 5 residues (adjust if needed)
components = [c for c in components if len(c) >= 5]
print(f"Components passing size filter (>=5): {len(components)}")

# ---- 4) Patch metrics
AROM = {"PHE","TYR","TRP"}
SMALL_ALI = {"ALA","VAL","LEU"}
POS, NEG = {"LYS","ARG","HIS"}, {"ASP","GLU"}
CHARGED = POS | NEG

def pca_thickness(residue_indices):
    ats = [ca_by_ridx.get(r) for r in residue_indices if r in ca_by_ridx]
    ats = [a for a in ats if a is not None]
    if not ats:
        return np.nan, np.nan, np.nan
    X = coords[ats] * 10.0  # Å
    Xc = X - X.mean(0)
    _, S, _ = np.linalg.svd(Xc, full_matrices=False)
    # S ~ extent along principal axes in Å
    s1, s2, s3 = (S.tolist() + [np.nan, np.nan, np.nan])[:3]
    return s1, s2, s3

def table_for(ridx_list):
    return tbl.set_index("ridx").loc[ridx_list].reset_index()

patch_rows, pymol_selections = [], []
for pid, comp in enumerate(components, start=1):
    sub = table_for(comp)
    nres = len(sub)
    area = float(sub["sasa_A2"].sum())
    s1, s2, s3 = pca_thickness(comp)
    planarity = s3 / (s1 + s2 + 1e-6) if np.all(np.isfinite([s1,s2,s3])) else np.nan
    arom_frac = (sub["resname"].isin(AROM)).mean()
    small_ali_frac = (sub["resname"].isin(SMALL_ALI)).mean()
    charged_frac = (sub["resname"].isin(CHARGED)).mean()
    mean_RSA = float(sub["RSA"].mean())
    # Heuristic epitope-like call
    epitope_like = (10 <= nres <= 25) and (200 <= area <= 800) and (np.isfinite(planarity) and planarity < 0.20)
    patch_rows.append({
        "patch_id": pid,
        "n_res": nres,
        "total_SASA_A2": area,
        "mean_RSA": mean_RSA,
        "s1_A": s1, "s2_A": s2, "s3_A(thickness)": s3,
        "planarity_score": planarity,
        "frac_aromatic": arom_frac,
        "frac_small_aliphatic": small_ali_frac,
        "frac_charged": charged_frac,
        "epitope_like_heuristic": epitope_like
    })
    sel = " or ".join(f"(chain {r.chain} and resi {int(r.resid)})" for _, r in sub.iterrows())
    pymol_selections.append((pid, sel))

# ---- 5) Save outputs safely (handles empty case)
if len(patch_rows) == 0:
    # Write an empty file with headers to keep your pipeline tidy
    cols = ["patch_id","n_res","total_SASA_A2","mean_RSA","s1_A","s2_A","s3_A(thickness)",
            "planarity_score","frac_aromatic","frac_small_aliphatic","frac_charged","epitope_like_heuristic"]
    patch_df = pd.DataFrame(columns=cols)
    patch_df.to_csv("SASA/epitope_patches.csv", index=False)
    with open("SASA/epitope_patches_pymol.txt","w") as fh:
        fh.write("# No patches found with current thresholds.\n")
    print("⚠️ No patches met the criteria. Consider relaxing thresholds: RSA≥0.25, size≥3, area 150–900 Å².")
else:
    patch_df = pd.DataFrame(patch_rows)
    patch_df = patch_df.sort_values(["epitope_like_heuristic","total_SASA_A2"], ascending=[False,False])
    patch_df.to_csv("SASA/epitope_patches.csv", index=False)
    with open("SASA/epitope_patches_pymol.txt","w") as fh:
        for pid, sel in pymol_selections:
            fh.write(f"sel patch{pid}, {sel}\nshow surface, patch{pid}\nset transparency, 0.25, patch{pid}\n\n")
    print(f"✅ Wrote SASA/epitope_patches.csv with {len(patch_df)} rows.")
    display(patch_df.head(8))
    print("PyMOL selections written to SASA/epitope_patches_pymol.txt")
```

Global Backbone Dihedral Statistics: </b> Across 263 residues in the NR1 ligand-binding domain (LBD), the average backbone deviation between the apo (4KCC) and agonist-bound (7EU8) structures was modest overall but punctuated by localized conformational changes: Mean |Δφ| = 20.8°, Mean |Δψ| = 23.3°, giving an overall Mean |Δ| = 22.0°. Most residues deviated less than 20° (Median |Δ| = 14.5°), but the upper 10% exceeded |Δ| = 43.9°. The maximum local backbone shift reached 141.9°, indicating a few residues undergo large torsional rearrangements during the transition from apo to bound. These values suggest that while most of the LBD backbone remains geometrically conserved upon ligand binding, a subset of residues experiences substantial local rotations that likely mediate hinge motion and cleft closure.
